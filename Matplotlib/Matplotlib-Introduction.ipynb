```python
import numpy as np
import matplotlib.pyplot as plt
```

## API

We will mostly start with `fig, ax = plt.subplots()` to generate a (empty) figure.

```python
fig, ax = plt.subplots()
```

Then we use `ax.plot()` to plot something.

```python
fig, ax = plt.subplots()
ax.plot([1,2,3,4])
```

```python
fig, ax = plt.subplots()
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
ax.plot(x, y)
```

It is then easy to plot a function by using `numpy.linspace`.

```python
fig, ax = plt.subplots()
x = np.linspace(-np.pi, +np.pi)
y = np.sin(x)
ax.plot(x, y)
```

## Older version

```python
x = np.linspace(-3, 3)
y = np.sin(x)
plt.plot(x, y)
```

## pyplot API

```python
plt.figure(figsize=(3, 2))
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.xlim(-4.0, 4.0)
plt.ylim(-1.0, 1.0)

plt.xticks(np.linspace(-4, 4, 9, endpoint=True))
plt.yticks(np.linspace(-1, 1, 5, endpoint=True))
plt.show()
```

## object-oriented API

```python
fig, ax = plt.subplots(figsize=(3, 2))
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)

ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))

ax.set_xlim(-4.0, 4.0)
ax.set_ylim(-1.0, 1.0)
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

x = np.linspace(-np.pi, +np.pi)

ax.plot(x, np.sin(x), 'ko-', label='sin(x)')
ax.plot(x, np.cos(x), 'bx--', label=r'$\cos(x)$')

ax.legend(fontsize=16)
ax.grid()
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('y=sin(x)')
```
