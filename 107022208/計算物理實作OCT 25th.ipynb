```python
import numpy as np
import matplotlib.pyplot as plt
x0 = 0.5
a = [0.8,1.2,2.8,3.2]
X = np.linspace(0,1,100)
for j in a:
    x = [x0]
    y = [0]
    Y = []

    for i in range(100):
        Y.append(j*X[i]*(1-X[i])) 
    for i in range(100):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))


    plt.plot(X,Y,label='f at a='+str(j))
    plt.plot(x,y,label='x_n at a='+str(j))
plt.plot(X,X)
plt.legend()
plt.show()

for j in a:
    x = [x0]
    y = [0]
    N = [0]
    xx = [x0] #use xx to represent x_n
    for i in range(100):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        N.append(i)
        xx.append(x[len(x)-1])

    plt.plot(N,xx,label='x_n v.s. n at a='+str(j))
plt.title('x_n v.s. n')
plt.legend()
plt.show()
```
