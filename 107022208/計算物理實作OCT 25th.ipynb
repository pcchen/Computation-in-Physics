## Introduction

Consider the logic eqatuion:
$$
  f(\alpha,x)=\alpha x(1-x).
$$

The iterative equation becomes
$$
  x_{n+1} = f(\alpha, x_n) = \alpha x_n (1-x_n).
$$

We will assume $x$ is restricted to the domain $(0,1)$.

### Tasks: Plot the function and find the attractor
* Set $\alpha=0.8$ and choose some initial point $x_0$, iterate $f$ and convince yourself that they all are attracted to zero. It might be helpful to plot $x_n$ v.s. $n$.
* Plot $f$ and the diagonal $y=x$ on the same plot. Are there any fixed-points other than $x=0$?
* Repeat for $\alpha=1.2$, $\alpha=2.8$, $\alpha=3.2$. What happens?

```python
import numpy as np
import matplotlib.pyplot as plt
x0 = 0.6
a = [0.8,1.2,2.8,3.2]
X = np.linspace(0,1,100)
plt.figure(figsize=(12,8))
for j in a:
    x = [x0]
    y = [0]
    Y = []
    for i in range(100):
        Y.append(j*X[i]*(1-X[i])) 
    for i in range(100):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))


    plt.plot(X,Y,label='f(x) at a='+str(j))
    plt.plot(x,y,label='x_n at a='+str(j))
plt.plot(X,X)
plt.legend()
plt.show()
plt.figure(figsize=(12,8))
for j in a:
    x = [x0]
    y = [0]
    N = [0]
    xx = [x0] #use xx to represent x_n
    for i in range(100):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        N.append(i)
        xx.append(x[len(x)-1])

    plt.plot(N,xx,".",label='x_n v.s. n at a='+str(j))
plt.title('x_n v.s. n')
plt.legend()
plt.show()
```

### Tasks: Trajectory of the attractor
* On the same graph, plot $f$, the diagonal $y=x$, and the trajectory $\{x_0, x_0\}$, $\{x_0, f(x_0)\}$, $\{f(x_0), f(x_0)\}$, $\{f(x_0), f(fx_0)\}$, $\dots$ for $\alpha=2.8, 3.2, 3.52, 3.54, 4.00$. Describe their differences.

```python
import numpy as np
import matplotlib.pyplot as plt
x0 = 0.6
a = [2.8,3.2,3.52,3.54,4.00]
X = np.linspace(0,1,100)
for j in a:
    plt.figure(figsize=(6,4))
    x = [x0]
    y = [0]
    Y = []
    for i in range(100):
        Y.append(j*X[i]*(1-X[i])) 
    for i in range(100):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))
    plt.plot(X,Y,label='f(x)')
    plt.plot(x,y,label='x_n')
    plt.plot(X,X,label='y=x')
    plt.title('a='+str(j))
    plt.legend()
    plt.show()
```

### Tasks: Bifucation diagram
* Plot the attractor as a function of $\alpha$ for $0<\alpha<4$.

```python
import numpy as np
import matplotlib.pyplot as plt
x0 = 0.6
a = np.linspace(0,4,4001)
alpha = []
xx=[]
plt.figure(figsize=(12,8))
for j in a:
    x = [x0]
    y = [0]
    for i in range(200):
        y.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        x.append(x[len(x)-1])
        y.append(y[len(y)-1])
        x.append(j*x[len(x)-1]*(1-x[len(x)-1]))
        if i > 180 :
            alpha.append(j)
            xx.append(x[len(x)-1])
plt.plot(alpha,xx,".")
plt.xlabel('alpha')
plt.ylabel('x_n')
plt.show()
```
