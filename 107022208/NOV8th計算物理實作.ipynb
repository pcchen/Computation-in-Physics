```python
import matplotlib.pyplot as plt
import numpy as np
v=[]
N = 100
X = range(N)
Y = range(N)
for i in range(N):
    v.append([])
    for j in range(N):
        v[i].append(0)
for k in range(1000):
    for i in range(N):
        for j in range(N):
            if i*j*(i-N+1)*(j-N+1) == 0:
                v[i][j] = 0
            elif i == N//2 and j == N//2:
                v[i][j] = 100
            else:
                v[i][j] = 0.25*(v[i][j-1]+v[i][j+1]+v[i+1][j]+v[i-1][j])

# for contour plot
fig, ax = plt.subplots(figsize=(10,8))
cs = ax.contour(X, Y, v)
cs = ax.contourf(X, Y, v)
fig.colorbar(cs)
plt.show()

fig, ax = plt.subplots(figsize=(10,8))
plt.plot(X,v[N//2])
plt.xlabel('x-position')
plt.ylabel('voltage')
plt.title('for y='+str(N//2))
plt.xticks(np.linspace(0,N,11))
plt.show()
```
