## I.Simple Harmonic Oscillation

###    (I)Calculation

For an object mass $m$, if it's subjected by a force field minus propotional to its displacement like

$F=-kx$

By Newton 2nd law $F=ma$ ,we got

$F=m\frac{d^2x}{dt^2}=-kx$

For initial condition $x(0)=x_0$,$v(0)=0$

If we solve the ODE(Ordinary Differiential Equation), we can get

$x=Asin(\omega t)+Bcos(\omega t)$ ,$\omega =\sqrt\frac k m$

Using the initial condition and derive

$x(t)=x_0cos(\omega t)$

The figure below is the relation between x-t

###    (II)About code

To verify the equation $x(t)=x_0cos(\omega t)$,we use 5 steps below to run the motion.

(1)The force will affect on the acceleration.

(2)The acceleration would affect the velocity by time.

(3)The velocity would affect on the position of the object by time. 

(4)For time $t$ pass $dt$,the $v$ will be changed by $adt$,the $x$ will be changed by $vdt$. 

(5)Record every value of $(x,t)$ into lists and plot it.

```python
import matplotlib.pyplot as plt
k=1 #force const
m=1 #mass
x=10 #amplitude
v=0 #velocity
a=0 #acceleration
t=0 #time
gamma=0 #no damped
dt=0.001 #delta time
X=[]
T=[]
for i in range(30000):
        X.append(x)
        T.append(t)
        a = -(k/m)*x #F=ma=-kx
        t += dt
        v += a*dt
        x += v*dt
plt.plot(T,X)
plt.xlabel('time(t)')
plt.ylabel('position(x)')

plt.show()
```

## II.Energy Conserved
To check if sum of the potential energy and the kinetic energy is constant, we use the same method to check.

Potential energy 

$=\frac{1}{2} kx^2$

Kinetic energy

$=\frac{1}{2} mv^2$

We want to verify

$\frac{1}{2} kx^2+\frac{1}{2} mv^2=const.$

This time we use 4 lists to plot the figure.

```python
import matplotlib.pyplot as plt
k=1 #force const
m=1 #mass
x=10 #amplitude
v=0 #velocity
a=0 #acceleration
t=0 #time
gamma=0 #no damped
dt=0.001 #delta time
T=[]
U=[]
K=[]
E=[]
for i in range(10000):
        Eu=0.5*k*x**2
        Ek=0.5*m*v**2
        Et=Eu+Ek
        T.append(t)
        U.append(Eu)
        K.append(Ek)
        E.append(Et)
        a = -(k/m)*x #F=ma=-kx
        t += dt
        v += a*dt
        x += v*dt
plt.plot(T,U,label="potential energy")
plt.plot(T,K,label="kinetic energy")
plt.plot(T,E,label="total energy")

plt.xlabel('Time(t)')
plt.ylabel('Energy(E)')


plt.legend()
plt.show()
```

## III.Phase Diagram

To know the relation between displacement(x) and velocity(v),we can draw a phase diagram to show the x-v figure.

Since the sum of the potential energy and the kinetic energy is constant,

$\frac{1}{2} kx^2+\frac{1}{2} mv^2=const.$

we can see the curve of x-v in SHM is ellipse.

Same method used by above but this time we use 6 lists to plot the phase diagram.

```python
import matplotlib.pyplot as plt
k=1 #force const
m=1 #mass
x1=10 #amplitude1
x2=20 #amplitude2
x3=30 #amplitude3
v1=0 #velocity1
v2=0 #velocity2
v3=0 #velocity3
a1=0 #acceleration1
a2=0 #acceleration2
a3=0 #acceleration3
t=0 #time
gamma=0 #no damped
dt=0.001 #delta time
X1=[]
X2=[]
X3=[]
V1=[]
V2=[]
V3=[]
for i in range(30000):
        X1.append(x1)
        V1.append(v1)
        X2.append(x2)
        V2.append(v2)
        X3.append(x3)
        V3.append(v3)
        a1 = -(k/m)*x1 #F=ma=-kx
        a2 = -(k/m)*x2
        a3 = -(k/m)*x3
        t += dt
        v1 += a1*dt
        v2 += a2*dt
        v3 += a3*dt
        x1 += v1*dt
        x2 += v2*dt
        x3 += v3*dt
plt.plot(X1,V1,label="x_0=10")
plt.plot(X2,V2,label="x_0=20")
plt.plot(X3,V3,label="x_0=30")

plt.xlabel('position(x)')
plt.ylabel('velocity(v)')

plt.legend()
plt.show()
```

## IV.Lissajous Curve

For 2-dimension SHM, we can use different initial phase,angular frequency at $x$-direction and $y$-direction.

We can get this curve by using

```python
import matplotlib.pyplot as plt
import numpy as np
from math import*
T=np.linspace(0,1000,10001)
X=[]
Y=[]

for i in range (10001):
        X.append(sin(T[i]))
        Y.append(sin(T[i]))

plt.plot(X,Y)
plt.xlabel('position(x)')
plt.ylabel('position(y)')
plt.title('omega1:1,phase=0')
plt.show()

X=[]
Y=[]

for i in range (10001):
        X.append(sin(2*T[i]))
        Y.append(sin(T[i]))

plt.plot(X,Y)
plt.xlabel('position(x)')
plt.ylabel('position(y)')
plt.title('omega2:1,phase=0')
plt.show()

X=[]
Y=[]

for i in range (10001):
        X.append(sin(T[i]+np.pi/2))
        Y.append(sin(T[i]))

plt.plot(X,Y)
plt.xlabel('position(x)')
plt.ylabel('position(y)')
plt.title('omega1:1,phase=+pi/2')
plt.show()

X=[]
Y=[]

for i in range (10001):
        X.append(sin(3*T[i]+np.pi/3))
        Y.append(sin(2*T[i]))

plt.plot(X,Y)
plt.xlabel('position(x)')
plt.ylabel('position(y)')
plt.title('omega3:2,phase=+pi/3')
plt.show()

X=[]
Y=[]

for i in range (10001):
        X.append(sin(8*T[i]+np.pi/4))
        Y.append(sin(3*T[i]))

plt.plot(X,Y)
plt.xlabel('position(x)')
plt.ylabel('position(y)')
plt.title('omega8:3,phase=+pi/4')
plt.show()
```

## V.Damping Oscillation

If we add an extra force, which is minus propotional to its velocity, the total energy will decrease by time. Assume the coefficient of the velocity decrease is $\gamma$ .It's amplitude and velocity both goes to zero by time. 
### (I)Calculation
For an object mass $m$, it's subjected by a force field like

$F=-kx-\gamma v$

This can be written by differiential equation below.

$F=m\frac{d^2x}{dt^2}=-kx-\gamma \frac {dx}{dt}$

By using $x(t)=x_0e^{\alpha t}$ and calaulate the ODE.

We can derive $\alpha= -\beta \pm \sqrt{\beta^2-\omega^2}$, which $\omega^2=\frac{k}{m}$,$\beta=\frac{\gamma}{2m}$

#### (1)For $\beta^2 < \omega^2$, $\sqrt{\beta^2-\omega^2}$ is imaginary number, the oscillation would be under-damping.



#### (2)For $\beta^2 = \omega^2$ ,$\sqrt{\beta^2-\omega^2}$ is zero, the oscillation would be critical-damping.



#### (3)For $\beta^2 > \omega^2$,$\sqrt{\beta^2-\omega^2}$ is real number, the oscillation would be over-damping.

To plot the 3 curves, we use same method in SHM. This time we add the extra force $-\gamma v$ to change the accerleration.

The figure below is the phase diagram.

```python
import matplotlib.pyplot as plt
k=1 #force const
m=1 #mass
x1=10 #amplitude1
x2=10 #amplitude2
x3=10 #amplitude3
v1=0 #velocity1
v2=0 #velocity2
v3=0 #velocity3
a1=0 #acceleration1
a2=0 #acceleration2
a3=0 #acceleration3
t=0 #time
gamma1=0.4 #under-damped
gamma2=2#critical-damped
gamma3=3.6#over-damped
dt=0.001 #delta time
X1=[]
X2=[]
X3=[]
T=[]
for i in range(30000):
        X1.append(x1)
        X2.append(x2)
        X3.append(x3)
        T.append(t)
        a1 = (-k*x1-gamma1*v1)/m #F=-kx-gamma*v
        a2 = (-k*x2-gamma2*v2)/m
        a3 = (-k*x3-gamma3*v3)/m
        t += dt
        v1 += a1*dt
        v2 += a2*dt
        v3 += a3*dt
        x1 += v1*dt
        x2 += v2*dt
        x3 += v3*dt
plt.plot(T,X1,label='under-damped')
plt.plot(T,X2,label='critical-damped')
plt.plot(T,X3,label='over-damped')

plt.xlabel('time(t)')
plt.ylabel('position(x)')

plt.legend()
plt.show()
```

### (II)Energy
To plot the kinetic energy and potential energy of damping oscillation. We use the code below.

```python
import matplotlib.pyplot as plt
k=1 #force const
m=1 #mass
x=10 #amplitude
v=0 #velocity
a=0 #acceleration
t=0 #time
gamma=0.4 #under damped
dt=0.001 #delta time
T=[]
U=[]
K=[]
E=[]
for i in range(10000):
        Eu=0.5*k*x**2
        Ek=0.5*m*v**2
        Et=Eu+Ek
        T.append(t)
        U.append(Eu)
        K.append(Ek)
        E.append(Et)
        a = -(k/m)*x-(gamma/m)*v #F=ma=-kx-gamma*v
        t += dt
        v += a*dt
        x += v*dt
plt.plot(T,U,label="potential energy")
plt.plot(T,K,label="kinetic energy")
plt.plot(T,E,label="total energy")

plt.xlabel('Time(t)')
plt.ylabel('Energy(E)')
plt.title('under-damped')


plt.legend()
plt.show()
```
