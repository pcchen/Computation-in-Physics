```python
import matplotlib.pyplot as plt
import numpy as np
v=[]
N = 100
X = range(N)
Y = range(N)
for i in range(N):
    v.append([])
    for j in range(N):
        v[i].append(0)
for k in range(4000):
    for i in range(N):
        for j in range(N):
            if i == 0: #*j*(j-N+1)
                v[i][j] = 100
            elif (i-N+1) == 0: #*j*(j-N+1)
                v[i][j] = -100
            elif j == N-1:
                v[i][j] = 0.25*(v[i][j-1]+v[i][0]+v[i+1][j]+v[i-1][j])
            #elif i == N//2 and j == N//2:
                #v[i][j] = 100
            else:
                v[i][j] = 0.25*(v[i][j-1]+v[i][j+1]+v[i+1][j]+v[i-1][j])

# for contour plot
fig, ax = plt.subplots(figsize=(10,8))
cs = ax.contour(X, Y, v)
cs = ax.contourf(X, Y, v)
fig.colorbar(cs)
#plt.show()

Ex = []
Ey = []
for i in range(N):
    Ex.append([])
    Ey.append([])
    for j in range(N):
        Ex[i].append(0)
        Ey[i].append(0)
for j in range(N):
    for i in range(N):
        if i == 0:
            Ey[i][j] = 0
        elif i == N-1:
            Ey[i][j] = 0
        else:
            Ey[i][j] = -0.5*(v[i+1][j]-v[i-1][j])
for i in range(N):
    for j in range(N):
        if j == 0:
            Ex[i][j] = -0.5*(v[i][j+1]-v[i][N-1])
        elif j == N-1:
            Ex[i][j] = -0.5*(v[i][0]-v[i][j-1])
        else:
            Ex[i][j] = -0.5*(v[i][j+1]-v[i][j-1])
X = np.array(X)
Y = np.array(Y)
Ex = np.array(Ex)
Ey = np.array(Ey)
# for electric field
ax.streamplot(X, Y, Ex, Ey, color='b')
ax.quiver(X, Y, Ex, Ey, color='b')
plt.show()
```
