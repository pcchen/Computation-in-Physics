```python
import numpy as np
import matplotlib.pyplot as plt
```

![simple harmonic oscillator](../計算物理/shm.png)

For an simple harmonic oscillator, we could write down the equation of motions as:
\begin{align}
\ddot{x}(t) + \omega_0^2 x(t)=0
\end{align}
where $\omega_0 = \sqrt{\frac{k}{m}}$

general solution for simple harmonic oscillator:
\begin{align}
x(t)=x_0\sin\left(\omega_0 t - φ\right)
\end{align}

```python
m = 1
k = 1
w = (k/m)**0.5

x0 = 1
phi = np.pi

t = np.linspace(0, 20, 2001)
x = x0*np.sin(w*t - phi)

plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
p = plt.gcf()
p.set_size_inches(7, 5)
plt.show()
```

## Kinetic energy, potential energy, and total energy

By differentiating the function of position, we have the function of velocity:
\begin{align}
v(t) = \frac{dx(t)}{dt} = \omega_0 x_0 \cos\left(\omega_0 t - φ\right)
\end{align}
And we can find the function of the kinetic energy and the potential energy. 

Kinetic energy:
\begin{align}
T = \frac12 m v^2
\end{align}
Potential energy:
\begin{align}
U = \frac12 k x^2
\end{align}

```python
v = w*x0*np.cos(w*t - phi)

T = 0.5*m*v**2
U = 0.5*k*x**2

E = T + U

plt.plot(t, T, 'r-', label = 'Kinetic energy')
plt.plot(t, U, 'b-', label = 'Potential energy')
plt.plot(t, E, 'g-', label = 'Total energy')

plt.legend(loc = 5)
plt.xlim(-1, 25)
plt.xlabel('t')
plt.ylabel('E(t)')
p = plt.gcf()
p.set_size_inches(12.5, 7)

plt.show()
```

## Phase diagram

Then, we can plot the $v(t)$ versus $x(t)$. By plotting phase diagram, we can figure out which state is this system in.

```python
x0 = [1, 1.5, 2, 2.5, 3]

for i in range(len(x0)):

    x = x0[i]*np.sin(w*t - phi)
    v = w*x0[i]*np.cos(w*t - phi)
    plt.plot(x, v, label = 'x0={}'.format(x0[i]))

plt.xlabel('x')
plt.ylabel('v')
plt.xlim(-3.5, 4.5)
plt.legend()
p = plt.gcf()
p.set_size_inches(7.5, 6)
plt.show()
```

## Two dimentional simple harmonic oscillation

Then, we consider the situation that the system is doing a 2-D simple harmonic motion. 

\begin{align}
x(t)=x_0\sin(\omega_xt-\phi_x)\\
y(t)=y_0\sin(\omega_yt-\phi_y)
\end{align}

In this section, we assume they have the same ammplitute and frequency, so we plot the motion of the oscillator with several phase differences.

```python
w = (k/m)**0.5
b = 0.1

x0 = 1
phi = np.array([0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2])*np.pi

fig, ax = plt.subplots(ncols=4, nrows=2, figsize=(4*5, 4*2))
ax = ax.flatten()

for i in range(8):

    x = x0*np.sin(w*t)
    y = x0*np.sin(w*t - phi[i])

    ax[i].plot(x, y)
    ax[i].set_xlim([-x0, x0])
    ax[i].set_ylim([-x0, x0])
    plt.xlabel('x')
    plt.ylabel('y')

plt.show()
```

### Lissajous curve

Then, we plot the 2-D simple harmonic motion, which the x and y direction have different elastic coefficiants. I set the $k_y$ values is two times and three times of $k_x$. Then, I plot Lissajous curves with several phase differences.

```python
wx = (k/m)**0.5
wy = 2*wx
wy2 = 3*wx
b = 0.1

x0 = 1
phi = np.array([0.75, 1, 1.25, 1.5])*np.pi

fig, ax = plt.subplots(ncols=4, nrows=2, figsize=(4*5, 4*2))
ax = ax.flatten()

for i in range(4):

    x = x0*np.sin(wx*t)
    y = x0*np.sin(wy*t - phi[i])

    ax[i].plot(x, y)
    ax[i].set_xlim([-x0, x0])
    ax[i].set_ylim([-x0, x0])
    plt.xlabel('x')
    plt.ylabel('y')
    
for i in range(4):

    x = x0*np.sin(wx*t)
    y = x0*np.sin(wy2*t - phi[i])

    ax[i+4].plot(x, y)
    ax[i+4].set_xlim([-x0, x0])
    ax[i+4].set_ylim([-x0, x0])
    plt.xlabel('x')
    plt.ylabel('y')

plt.show()
```
