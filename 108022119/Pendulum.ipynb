```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = 1
g = 9.8
theta0 = 2*np.pi/3
T = 2*np.pi*np.sqrt(l/g)
w = 2*np.pi/T

dt = [0.1, 0.01, 0.001]
```

## Euler method

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(4*5, 4*5))
ax = ax.flatten()

k = 0

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:
        
        dtheta.append(dtheta[j] + -w**2*np.sin(theta[j])*dt[i])
        theta.append(theta[j] + dtheta[j]*dt[i])
        
        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
    
    x = l*np.sin(theta)
    y = -l*np.cos(theta)
    
    ax[2*k].plot(x, y)
    ax[2*k+1].plot(t, theta)
    k = k+1


plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(4*5, 4*2))
ax = ax.flatten()

last_pt = []

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:
        
        dtheta.append(dtheta[j] - w**2*np.sin(theta[j])*dt[i])
        theta.append(theta[j] + dtheta[j]*dt[i])
        
        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
    
    x = l*np.sin(theta)
    y = -l*np.cos(theta)
    
    last_pt.append(theta[-1])
    
    ax[0].plot(x, y)
    ax[1].plot(t,theta)

plt.show()
```

```python
xlpt = l*np.sin(last_pt)
ylpt = -l*np.cos(last_pt)

for i in range(len(dt)):

    plt.plot(xlpt[i],ylpt[i],'o',label = 'dt={}'.format(dt[i]))

p = plt.gcf()
p.set_size_inches(7, 5)
plt.legend()
```

## Staggered algorithm

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(4*5, 4*5))
ax = ax.flatten()

k = 0

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:
        dtheta.append(dtheta[j] + -w**2*np.sin(theta[j])*dt[i])
        theta.append(theta[j] + (dtheta[j+1]+dtheta[j])*dt[i]/2)

        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
    
    xx = l*np.sin(theta)
    yy = -l*np.cos(theta)
    
    ax[2*k].plot(xx, yy)
    ax[2*k+1].plot(t, theta)
    k = k + 1


plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(4*5, 4*2))
ax = ax.flatten()

last_pt = []

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:
        
        dtheta.append(dtheta[j] + -w**2*np.sin(theta[j])*dt[i])
        theta.append(theta[j] + (dtheta[j+1]+dtheta[j])*dt[i]/2)
        
        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
    
    x = l*np.sin(theta)
    y = -l*np.cos(theta)
    
    last_pt.append(theta[-1])
    
    ax[0].plot(x, y)
    ax[1].plot(t,theta)

plt.show()
```

```python
xlpt = l*np.sin(last_pt)
ylpt = -l*np.cos(last_pt)

for i in range(len(dt)):

    plt.plot(xlpt[i],ylpt[i],'o',label = 'dt={}'.format(dt[i]))

p = plt.gcf()
p.set_size_inches(7, 5)
plt.legend()
```

# Velocity Verlet algorithm

```python
fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(4*5, 4*5))
ax = ax.flatten()

k = 0

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:

        dtheta.append(dtheta[2*j] - w**2*np.sin(theta[j])*dt[i]/2)
        theta.append(theta[j] + dtheta[2*j+1]*dt[i])
        dtheta.append(dtheta[2*j+1] - w**2*np.sin(theta[j+1])*dt[i]/2)

        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
            
    xx = l*np.sin(theta)
    yy = -l*np.cos(theta)
    
    ax[2*k].plot(xx, yy)
    ax[2*k+1].plot(t, theta)
    k = k + 1


plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(4*5, 4*2))
ax = ax.flatten()

last_pt = []

for i in range(len(dt)):
    
    theta = []
    dtheta = []
    t = []
    
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)

    ti = 0
    j = 0
    
    while ti <= 2*T:
        
        dtheta.append(dtheta[2*j] - w**2*np.sin(theta[j])*dt[i]/2)
        theta.append(theta[j] + dtheta[2*j+1]*dt[i])
        dtheta.append(dtheta[2*j+1] - w**2*np.sin(theta[j+1])*dt[i]/2)
        
        j = j + 1
        ti = ti + dt[i]
        t.append(ti)
    
    x = l*np.sin(theta)
    y = -l*np.cos(theta)
    
    last_pt.append(theta[-1])
    
    ax[0].plot(x, y)
    ax[1].plot(t,theta)

plt.show()
```
