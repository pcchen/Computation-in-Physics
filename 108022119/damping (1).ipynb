```python
import numpy as np
import matplotlib.pyplot as plt
```

The equation of motions is:

$\ddot{x}(t) + 2\beta \dot{x}(t) + \omega_0^2 x(t)=0$

general solution for damped oscillator:

$x(t)=e^{-\beta t}\left(A_1e^{\sqrt{\beta^2-\omega_0^2}\;t+i\phi} + A_2e^{-\sqrt{\beta^2-\omega_0^2}\;t-i\phi}\right)$

Underdamping:

$\beta^2 < \omega_0^2$

Overdamping:

$\beta^2 > \omega_0^2$

Critically damping:

$\beta^2 = \omega_0^2$






```python
m = 1
k = 1
w = (k/m)**0.5
b = 0.2

x0 = 1
phi = np.pi

t = np.linspace(0, 20, 2001)

if b**2 > w**2:
    
    w1 = (b**2 - w**2)**0.5
    A1 = 0.5*x0*(1 + b/w1)
    A2 = 0.5*x0*(1 - b/w1)
    x = np.exp(-b*t)*(A1*np.exp(w1*t) + A2*np.exp(-w1*t))*np.cos(phi)
    v = -b*x + w1*np.exp(-b*t)*(A1*np.exp(w1*t) - A2*np.exp(-w1*t))*np.cos(phi)
    
elif b**2 < w**2:
    
    x = x0*np.exp(-b*t)*np.cos((w**2 - b**2)**0.5*t + phi)
    v = -b*x - (w**2 - b**2)**0.5*x0*np.exp(-b*t)*np.sin((w**2 - b**2)**0.5*t + phi)
    
else:
    
    x = (1 + b*t)*x0*np.exp(-b*t)*np.cos(phi)
    v = -b*x + b*x0*np.exp(-b*t)*np.cos(phi)

plt.plot(t, x)
plt.xlabel('t')
plt.ylabel('x(t)')
p = plt.gcf()
p.set_size_inches(7, 5)
plt.show()
```

## Kinetic energy, potential energy, and total energy

```python
T = 0.5*m*v**2
U = 0.5*k*x**2

E = T + U

plt.plot(t, T, 'r-', label = 'Kinetic energy')
plt.plot(t, U, 'b-', label = 'Potential energy')
plt.plot(t, E, 'g-', label = 'Total energy')

plt.legend(loc = 5)
plt.xlim(-1, 22)
p = plt.gcf()
p.set_size_inches(12.5, 7)
plt.show()
```

## Phase diagram
### Underdamping

```python
m = 1
k = 1
w = (k/m)**0.5
b = 0.1

x0 = 1
phi = np.pi

t = np.linspace(0, 20, 2001)

x = x0*np.exp(-b*t)*np.cos((w**2 - b**2)**0.5*t + phi)
v = -b*x - (w**2 - b**2)**0.5*x0*np.exp(-b*t)*np.sin((w**2 - b**2)**0.5*t + phi)

plt.plot(x, v, label = 'x0={}'.format(x0))

plt.xlabel('x')
plt.ylabel('v')
plt.legend()
p = plt.gcf()
p.set_size_inches(7.5, 6)
plt.show()
```

### Overdamping

```python
m = 1
k = 1
w = (k/m)**0.5
b = 2

x0 = 1
phi = np.array([1, 2])*np.pi

t = np.linspace(0, 20, 2001)

for i in range(len(phi)):
    
    w1 = (b**2 - w**2)**0.5
    A1 = 0.5*x0*(1 + b/w1)
    A2 = 0.5*x0*(1 - b/w1)
    x = np.exp(-b*t)*(A1*np.exp(w1*t) + A2*np.exp(-w1*t))*np.cos(phi[i])
    v = -b*x + w1*np.exp(-b*t)*(A1*np.exp(w1*t) - A2*np.exp(-w1*t))*np.cos(phi[i])
    plt.plot(x, v, label = 'phi={}'.format(phi[i]))

plt.xlabel('x')
plt.ylabel('v')
plt.legend()
p = plt.gcf()
p.set_size_inches(7.5, 6)
plt.show()
```

### Critically damping

```python
m = 1
k = 1
w = (k/m)**0.5
b = 1

x0 = 1
phi = np.array([0, 1])*np.pi

t = np.linspace(0, 20, 2001)

for i in range(len(phi)):
    
    x = (1 + b*t)*x0*np.exp(-b*t)*np.cos(phi[i])
    v = -b*x + b*x0*np.exp(-b*t)*np.cos(phi[i])
    plt.plot(x, v, label = 'phi={}'.format(phi[i]))

plt.xlabel('x')
plt.ylabel('v')
plt.legend()
p = plt.gcf()
p.set_size_inches(7.5, 6)
plt.show()
```
