```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

### 1D Jacobi and Gauss 

```python
n = 101

x = np.linspace(0, 1, n)
y1 = np.zeros(n)
y2 = np.zeros(n)

y1[-1] = 10
y2[-1] = 10

N = 102

plt.plot(x, 10*x)
p = plt.gcf()
p.set_size_inches(10, 8)

for i in range(N):
    
    if i%10 == 1:
        
        plt.plot(x, y1, 'o-', label = 'iteration =' + str(i))
        
    for j in range(n-2):
        
        y1[j+1] = 0.5*(y1[j] + y1[j+2])
        
plt.legend()
plt.show()


plt.plot(x, 10*x)
p = plt.gcf()
p.set_size_inches(10, 8)

for i in range(N):
    
    if i%10 == 1:
        
        plt.plot(x, y2, 'o-', label = 'iteration =' + str(i))
        
    for j in range(n-2):
                
        y2[n-j-2] = 0.5*(y2[n-j-1] + y2[n-j-3])
        
plt.legend()
plt.show()
```

### 2D Jacobi and Gauss

```python
n = 0
nmax = 10000

h = 0.05

Lx = 10
Ly = 10

nx = int(Lx/h) + 1
ny = int(Ly/h) + 1

x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)

v = np.zeros([nx, ny])
v[-1, :] = 1
v[0, :] = 1
v[:, -1] = 1
v[:, 0] = 1

v1 = v

while n <= nmax:

    v1[1:nx-1, 1:ny-1] = 0.25*(v[0:nx-2, 0:ny-2] + v[2:nx, 0:ny-2] + v[0:nx-2, 2:ny] + v[2:nx, 2:ny])
        
    n += 1
    
v = v1
    
plt.contourf(x, y, v)
plt.colorbar()

plt.show()

plt.show()
```

```python
def twodgauss(x, y, v):

    nx = len(x)
    ny = len(y)
    
    n = 0
    nmax = 10000
    
    while n <= nmax:

        v[1:nx-1, 1:ny-1] = 0.25*(v[0:nx-2, 0:ny-2] + v[2:nx, 0:ny-2] + v[0:nx-2, 2:ny] + v[2:nx, 2:ny])

        n += 1

    plt.contourf(x, y, v)
    plt.colorbar()

    plt.show()

    plt.show()

    Ex,Ey = np.gradient(v)
    fig,ay = plt.subplots(ncols = 2,nrows = 1,figsize = (16,8))
    ay[0].streamplot(y, x, -Ey, -Ex, color='k')
    ay[1].quiver(y, x, -Ey, -Ex, color = 'k')

Lx = 10
Ly = 10

nx = 101
ny = 101
    
x = np.linspace(0, Lx, nx)
y = np.linspace(0, Ly, ny)

v = np.zeros([nx, ny])
v[-1, :] = 10
v[0, :] = 10
v[:, -1] = 10
v[:, 0] = 10
    
twodgauss(x, y, v)
```

```python
def relaxation(Nx, Ny, nx, ny):

    nmax = 10

    x = np.linspace(0, Nx, nx+1)
    y = np.linspace(0, Ny, ny+1)

    v = np.zeros([nx + 1, ny + 1])

    v[:, 0] = 10
    v[0, :] = 10
    v[nx,:] = 10
    v[:,ny] = 10
    
    v1 = v

    for n in range(nmax):

        v1[1:nx-1, 1:ny-1] = 0.25*(v[0:nx-2, 0:ny-2] + v[2:nx, 0:ny-2] + v[0:nx-2, 2:ny] + v[2:nx, 2:ny])

    v = v1
```
