```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(a, x):
    
    y = a*x*(1-x)
    
    return y
```

```python
a = 0.8
x0 = 0.5
nmax = 100


xn = np.zeros(nmax+1)
n = np.zeros(nmax+1)

xn[0] = f(a, x0)

for i in range(nmax):

    xn[i+1] = f(a, xn[i])
    n[i+1] = i+1

plt.plot(n, xn)
plt.show()

x = np.linspace(0, 1, 100)
y = f(a, x)
yl = x
    
plt.plot(x, y)
plt.plot(x, yl)
plt.show()
```

```python
a = [2.80, 3.20, 3.52, 3.54, 4.00]

x1 = np.zeros(200)
y1 = np.zeros(200)
x1[0] = 0.2

fig, ax = plt.subplots(ncols=3, nrows=2, figsize=(5*3, 4*2))
ax = ax.flatten()

for j in range(len(a)):
    
    y = f(a[j], x)

    for i in range(len(x1)-1):

        if i %2 == 0:

            x1[i+1] = x1[i]
            y1[i+1] = f(a[j], x1[i])

        else:

            x1[i+1] = y1[i]
            y1[i+1] = y1[i]

    ax[j].plot(x, y)
    ax[j].plot(x, yl)

    for i in range(len(x1) - 1):

        ax[j].arrow(x1[i], y1[i], x1[i+1]-x1[i], y1[i+1]-y1[i], length_includes_head = 'true', head_width = 0.015, width = 0.00005)

plt.show()
```

```python
a = np.zeros(50000)
xn = np.zeros(50000)
ai = 2.8

for i in range(len(a)):
    
    a[i] = ai
    
    if i % 200 == 199:
        
        ai += 0.005
      
i = 0

while i < len(a):
    
    x1 = 0.2
    y1 = 0
    
    for j in range(10000):
    
        if j % 2 == 0:
        
            x1 = x1
            y1 = f(a[i], x1)

        else:
        
            x1 = y1
            y1 = y1
    
        if j >= 9800:
            
            xn[i] = y1
            i = i+1
            
p = plt.gcf()
plt.plot(a, xn, 'bo', markersize = 1.5)
p.set_size_inches(10, 8)
plt.show()
```
