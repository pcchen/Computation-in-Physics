```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
def f(a, x):
    
    y = a*x*(1-x)
    
    return y

def df(a, x):
    
    y = a*(1 - 2*x)

    return y
```

```python
a = [0.8, 1.2, 2.8, 3.2]
x0 = 0.5
nmax = 100

fig, ax = plt.subplots(ncols=4, nrows=2, figsize=(5*4, 4*2))
ax = ax.flatten()

xn = np.zeros(nmax+1)
n = np.zeros(nmax+1)

for j in range(len(a)):

    xn[0] = f(a[j], x0)

    for i in range(nmax):

        xn[i+1] = f(a[j], xn[i])
        n[i+1] = i+1

    ax[j].plot(n, xn, 'o', markersize = 1.2)

    x = np.linspace(0, 1, 100)
    y = f(a[j], x)
    yl = x

    ax[j+4].plot(x, y)
    ax[j+4].plot(x, yl)
    
    
plt.show()
```
