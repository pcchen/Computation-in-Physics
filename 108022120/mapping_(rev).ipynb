## Mapping

$$
x_{n+1} = f(\alpha , x_n)
$$

$$
x_{n+1} = \alpha x_n (1-x_n)
$$

```python
x0 = 0.1
alpha = 2
def f(X):
    xx = alpha*X*(1-X)
    return xx
def ff(X):
    xx = X
    return X
x = [0]
t = [x0]
for i in range(100):
    t.append(t[2*i])
    x.append(f(t[2*i]))
    t.append(x[2*i+1])
    x.append(x[2*i+1])

#plt.plot(t,x)

fig, ax = plt.subplots(figsize = (10,8))

t1 = np.linspace(0,1)
x1 = f(t1)
x2 = t1
ax.plot(t1,x1)
ax.plot(t1,x2)
ax.plot(t,x)
plt.grid()
```

## superstable
### f(x) = x

```python
def f(X,A):
    tem_1 = A*X*(1-X)
    return tem_1
    
def f2(X,N,A):
    tem_2 = X
    for i in range(N):
        tem_2 = f(tem_2,A)
    return tem_2-0.5

alpha = np.linspace(0,4,40001)
x1 = f2(0.5,1,alpha)
x2 = f2(0.5,2,alpha)
x4 = f2(0.5,4,alpha)
x8 = f2(0.5,8,alpha)
x16 = f2(0.5,16,alpha)
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(alpha,x1,'.',markersize = 0.2)
ax.plot(alpha,x2,'.',markersize = 0.2)
ax.plot(alpha,x4,'.',markersize = 0.2)
ax.plot(alpha,x8,'.',markersize = 0.2)
ax.plot(alpha,x16,'.',markersize = 0.2)
yyy = np.zeros(40001)
ax.plot(alpha,yyy)
ax.grid()
```

# Find root and Find $\delta_n$

```python
R = np.zeros((5,5))
#x1
def fr(X):
    return f2(0.5,1,X)
root = optimize.brentq(fr, 1.8, 2.2)
R[0][0] = root
#x2
def fr(X):
    return f2(0.5,2,X)
root = optimize.brentq(fr, 1.8, 2.2)
R[1][0] = root

root = optimize.brentq(fr, 3, 3.5)
R[1][1] = root

#x4
def fr(X):
    return f2(0.5,4,X)
root = optimize.brentq(fr, 1.8, 2.2)
R[2][0] = root

root = optimize.brentq(fr, 3, 3.4)
R[2][1] = root

root = optimize.brentq(fr, 3.4, 3.55)
R[2][2] = root

#x8
def fr(X):
    return f2(0.5,8,X)
root = optimize.brentq(fr, 1.8, 2.2)
R[3][0] = root

root = optimize.brentq(fr, 3, 3.4)
R[3][1] = root

root = optimize.brentq(fr, 3.4, 3.5)
R[3][2] = root

root = optimize.brentq(fr, 3.5, 3.56)
R[3][3] = root

#x16
def fr(X):
    return f2(0.5,8,X)
root = optimize.brentq(fr, 1.8, 2.2)
R[4][0] = root

root = optimize.brentq(fr, 3, 3.4)
R[4][1] = root

root = optimize.brentq(fr, 3.4, 3.5)
R[4][2] = root

root = optimize.brentq(fr, 3.5, 3.56)
R[4][3] = root

root = optimize.brentq(fr, 3.56, 3.67)
R[4][4] = root

print(R)
```

## $\delta_n = \frac{\alpha_n - \alpha_{n-1}}{\alpha_{n+1} - \alpha_n}$

```python
def de(X):
    return (R[4][X] - R[4][X-1])/(R[4][X+1]-R[4][X])

print(de(1),de(2),de(3))
```

## Lyapunov exponents

```python
def f(X):
    xx = alpha*X*(1-X)
    return xx
N1 = 10000
lamb = []
for alpha in [np.linspace(0,4,1201)]:
    tk = [0.4]
    xk = [0]
    for i in range(N1):
        tk.append(tk[2*i])
        xk.append(f(tk[2*i]))
        tk.append(xk[2*i+1])
        xk.append(xk[2*i+1])
    tt = np.log(abs(0.4))/N1
    for i in range(2*N1-1):
        tt = tt + np.log(abs(alpha-2*alpha*tk[i+1]))/N1
    lamb.append(tt)
print (lamb[0][1:])
print(alpha[1:])
```

```python
fig, ax = plt.subplots()
ax.plot(alpha[1:],lamb[0][1:])
t = np.linspace(0,4.0,50)
x = np.zeros(50)
ax.plot(t,x,'.')
```
