# Euler method

```python
L = 1
g = 9.8
theta0 = 2*np.pi/3
period = 2*np.pi*np.sqrt(L/g)
omega = np.sqrt(g/L)

def ddtheta(x):
    
    a = -omega**2*np.sin(x)
    return a

fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(4*5, 4*2))
ax = ax.flatten()

j = 0
for dt in [0.1,0.01,0.001]:
    i = 0
    theta = []
    dtheta = []
    t = []
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)
    T = 0
    while T <= period*50/101:
        dtheta.append(dtheta[i] + ddtheta(theta[i])*dt)
        theta.append(theta[i] + dtheta[i]*dt)
        i = i + 1
        T = T + dt
        t.append(T)
    
    xx = L*np.sin(theta)
    yy = -L*np.cos(theta)
    
    ax[2*j].plot(xx, yy)
    ax[2*j+1].plot(t, theta)
    j = j+1


plt.show()
```

```python
fig, az = plt.subplots(figsize=(4*5, 4*2))

xp = L*np.sin(LsPt)
yp = -L*np.cos(LsPt)

az.plot(xp[0],yp[0],'o',label = '0.1')
az.plot(xp[1],yp[1],'o',label = '0.01')
az.plot(xp[2],yp[2],'o',label = '0.001')
az.legend()
```

# pendulum -- period
## Euler method

```python
L = 1
g = 9.8
theta0 = 2*np.pi/3
period = 2*np.pi*np.sqrt(L/g)
omega = np.sqrt(g/L)

def ddtheta(x):
    
    a = -omega**2*np.sin(x)
    return a

fig, ax = plt.subplots(ncols=2, nrows=3, figsize=(4*5, 4*2))
ax = ax.flatten()

bp = []

j = 0
for dt in [0.1,0.01,0.001]:
    i = 0
    theta = []
    dtheta = []
    t = []
    theta.append(theta0)
    dtheta.append(0)
    t.append(0)
    T = 0
    while T <= period:
        dtheta.append(dtheta[i] + ddtheta(theta[i])*dt)
        theta.append(theta[i] + dtheta[i]*dt)
        i = i + 1
        T = T + dt
        t.append(T)
        if dtheta[i] > 0:
            break
    
    bp.append(t[-2]*2)
        
    
    xx = L*np.sin(theta)
    yy = -L*np.cos(theta)
    
    ax[2*j].plot(xx, yy)
    ax[2*j+1].plot(t, theta)
    j = j+1

plt.show()
print(bp)
print(period)
```
