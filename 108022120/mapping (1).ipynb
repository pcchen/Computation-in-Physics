```python
import numpy as np
import matplotlib.pyplot as plt
```

## Mapping

$$
x_{n+1} = f(\alpha , x_n)
$$

$$
x_{n+1} = \alpha x_n (1-x_n)
$$

```python
x0 = 0.1
alpha = 2
def f(X):
    xx = alpha*X*(1-X)
    return xx
def ff(X):
    xx = X
    return X
x = [0]
t = [x0]
for i in range(100):
    t.append(t[2*i])
    x.append(f(t[2*i]))
    t.append(x[2*i+1])
    x.append(x[2*i+1])

#plt.plot(t,x)

fig, ax = plt.subplots(figsize = (10,8))

t1 = np.linspace(0,1)
x1 = f(t1)
x2 = t1
ax.plot(t1,x1)
ax.plot(t1,x2)
ax.plot(t,x)
plt.grid()
```
