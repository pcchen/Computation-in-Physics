## Simple Harmonic Oscillator

By substituting the Hooke's Law $F(x)=-kx$ into the Newtonian eauation $F=m\ddot{x}$ obtain the equation of the motion

$$
    \ddot{x}(t) + \omega^2_0 x(t) =0,
$$

where $\omega^2_0 \equiv k/m$. The solution is of the form

$$
    x(t) = A \sin(\omega_0 t - \delta),
$$

were $A$ and $\delta$ are determined by the initial conditions.

```python
m = 1
k = 1

A = 1
delta = -np.pi/4

t = []
x = []

def X(T):
    xx = A*np.sin(np.sqrt(k/m)*T-delta)
    return xx

for i in range (0,1000):
    t.append(i/100)
    x.append(X(t[i]))

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, x)
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.grid()
```

## Energy

From $x(t) = A \sin(\omega_0 t - \delta)$, we can obtain the velocity $\dot{x}(t) = v(t) = \omega_0 A \cos(\omega_0 t - \delta)$.

The kinetic energy

$$
    K = \frac{1}{2} m \dot{x}^2(t) = \frac{1}{2} m \omega_0^2 A^2 \cos^2(\omega_0 t - \delta) = \frac{1}{2} k A^2 \cos^2(\omega_0 t - \delta).
$$

The potential energy

$$
    U = \frac{1}{2} k x^2(t) = \frac{1}{2} k A^2 \sin^2(\omega_0 t - \delta).
$$

The total energy

$$
    E = K + U = \frac{1}{2} k A^2.
$$

```python
m = 1
k = 1
c = 1
A = 1

delta = -np.pi/4
zeta = c/(2*np.sqrt(m*k))
omega = np.sqrt(k/m)

t = []
x = []
v = []
K = []
U = []
E = []

def X(T):
    xx = A*np.sin(np.sqrt(k/m)*T-delta)
    return xx
def V(T):
    vv = np.sqrt(k/m)*A*np.cos(np.sqrt(k/m)*T-delta)
    return vv

for i in range (0,1500):
    t.append(i/100)
    x.append(X(t[i]))
    v.append(V(t[i]))
    K.append(0.5*m*v[i]*v[i])
    U.append(0.5*k*x[i]*x[i])
    E.append(K[i]+U[i])

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, K, label='Kinetic energy')
ax.plot(t, U, label='Potential energy')
ax.plot(t, E, 'k--', label='Total energy')

ax.set_xlabel('t')
ax.legend()
```

## Simple Harmonic Oscillations in Two Dimensions

Consider the harmonic oscillation in 2D, with the equation $F_x = -kx$ and $F_y = -ky$. The equations of motions are

$$
    \ddot{x}(t) + \omega_0^2 x(t) = 0,
$$
$$
    \ddot{y}(t) + \omega_0^2 y(t) = 0.
$$

The solutions are

$$
    x(t) = A \cos(\omega_0 t - \alpha),
$$
$$
    y(t) = B \cos(\omega_0 t - \beta),
$$

where $A$, $B$, $\alpha$ and $\beta$ are determined by the initial conditions.

```python
k = 1
m = 1
omega = np.sqrt(k/m)
A = 1
B = 1
delta = np.pi/4
alpha = np.pi/2
beta = alpha + delta

t = []
x = []
y = []

def X(T):
    xx = A*np.cos(omega*T - alpha)
    return xx

def Y(T):
    yy = B*np.cos(omega*T - beta)
    return yy

for i in range(1000):
    t.append(i/100)
    x.append(X(t[i]))
    y.append(Y(t[i]))

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y)
ax.plot([-A, +A], [0, 0], 'k')
ax.plot([0, 0], [-B, +B], 'k')
ax.set_xlim([-A, +A])
ax.set_ylim([-B, +B])
```

We can use different $\delta$ to plot different pscillations.

```python
k = 1.
m = 1.
omega = np.sqrt(k/m)
A = 1.
B = 1.

alpha = 0

beta = 0.
delta = np.array([90, 120, 150, 180, 210, 240, 270, 300, 330, 360])/180.*np.pi

def X(T):
    xx = A*np.cos(omega*T - alpha)
    return xx

def Y(T):
    yy = B*np.cos(omega*T - beta)
    return yy

fig, ax = plt.subplots(ncols=5, nrows=2, figsize=(4*5, 4*2))
ax = ax.flatten()

for j in range(10):
    alpha = beta + delta[j]
    t = []
    x = []
    y = []
    for i in range(1000):
        t.append(i/100)
        x.append(X(t[i]))
        y.append(Y(t[i]))
    ax[j].plot(x, y)
    ax[j].plot([-A, +A], [0, 0], 'k')
    ax[j].plot([0, 0], [-B, +B], 'k')
    ax[j].set_xlim([-A, +A])
    ax[j].set_ylim([-B, +B])
```

## Lissajour curve

In the general case, $k_x \neq k_y$ and $ \omega_x = \sqrt{\frac{k_x}{m}} \neq \omega_y = \sqrt{\frac{k_y}{m}}$

The equations of motions become

$$
  \ddot{x}(t)+\omega^2_x x(t) = 0,
$$
$$
  \ddot{y}(t)+\omega^2_y y(t) = 0,
$$

The solutions become

$$
  x(t) = A \cos(\omega_x t - \alpha)
$$
$$
  y(t) = B \cos(\omega_y t - \beta)
$$

where $A$, $B$, and $\delta$ are determined by the initial conditions.

The path of the motion is no longer an ellipse but a **Lissajour curve**.

```python
A = 1.
B = 1.
omegax = 1.
omegay = 2*omegax

alpha = 0
beta = np.pi*3/4
delta = np.array([0., np.pi/3, np.pi/2])

def X(T):
    xx = A*np.cos(omegax*T - alpha)
    return xx

def Y(T):
    yy = B*np.cos(omegay*T - beta)
    return yy

fig, ax = plt.subplots(ncols=3, figsize=(6*3, 6))
ax = ax.flatten()

for i in range(3):
    alpha = beta + delta[i]
    t = []
    x = []
    y = []
    for j in range(1000):
        t.append(j/100)
        x.append(X(t[j]))
        y.append(Y(t[j]))
    ax[i].plot(x, y)
    ax[i].plot([-A, +A], [0, 0], 'k')
    ax[i].plot([0, 0], [-B, +B], 'k')
    ax[i].set_xlim([-A, +A])
    ax[i].set_ylim([-B, +B])
```

### Phase diagrams

We may consider the quantities $(x(t), \dot{x}(t))$ to be the coordinates of a point in a two-dimensional "phase space".

```python
k = 1.
m = 1.
omega = np.sqrt(k/m)
delta = np.pi/4.

def X(T):
    xx = A*np.sin(np.sqrt(k/m)*T-delta)
    return xx
def V(T):
    vv = np.sqrt(k/m)*A*np.cos(np.sqrt(k/m)*T-delta)
    return vv

fig, ax = plt.subplots(figsize=(10,8))
for A in [0.5, 1.0, 1.5, 2.0]:
    t = []
    x = []
    v = []
    for j in range (1000):
        t.append(j/100)
        x.append(X(t[j]))
        v.append(V(t[j]))
    ax.plot(x, v, 'k')

ax.set_xlabel('x', fontsize=24)
ax.set_ylabel('v', fontsize=24)
ax.grid()
```
