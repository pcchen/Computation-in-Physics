```python
import numpy as np
import matplotlib.pyplot as plt
```

## gauss

```python
cut = 101
a = np.linspace(0,1,cut)
ini = 0.5*np.zeros(cut)
ini[0] = 0
ini[-1] = 1
N = 100
def f(X):
    return X
fig, ax = plt.subplots(figsize = (10,8))
for i in range(N):
    if i%10 == 1:
        ax.plot(a,ini,'o-',markersize = 5,label = 'iteration =' + str(i))
        ax.plot(a,f(a),'k-')
        
    for j in range(len(ini)-2):
        k = len(ini)-j-2
        ini[k] = (ini[k-1]+ini[k+1])/2
plt.legend()
```

## jacobi

```python
cut = 101
a = np.linspace(0,1,cut)
ini = np.zeros(cut)
ini[0] = 0
ini[-1] = 1
N = 100
fig, ax = plt.subplots(figsize = (10,8))
for i in range(N):
    if i%10 == 1:
        ax.plot(a,ini,'o-',markersize = 5,label = 'iteration =' + str(i))
        ax.plot(a,f(a),'k-',markersize = 5)
        
    for j in range(len(ini)-2):
        k = j+1
        ini[k] = (ini[k-1]+ini[k+1])/2
plt.legend()
```

## 2D--Gauss

```python
length = 11
width = 10
N = 100
cut = 5
x = np.linspace(0,length,length*cut+1)
y = np.linspace(0,width,width*cut+1)
X,Y = np.meshgrid(x,y,indexing = 'ij')
V = np.zeros((cut*length+1,cut*width+1))
V[0,:] = 10
V[:,0] = 10
V[length*cut,:] = 10
V[:,width*cut] = 10
fig, ax = plt.subplots(figsize = (10,8))
for n in range(N):
    for i in range(cut*length+1):
        for j in range(cut*width+1):
            if i != 0 and i != cut*length:
                if j != 0 and j != cut*width:
                    V[i][j] = (V[i-1][j]+V[i][j-1]+V[i+1][j]+V[i][j+1])/4


cs = ax.contourf(X,Y,V)
fig.colorbar(cs)
```

## 2D -- jocobi

```python
length = 11
width = 10
N = 100
cut = 5
x = np.linspace(0,length,length*cut+1)
y = np.linspace(0,width,width*cut+1)
X,Y = np.meshgrid(x,y,indexing = 'ij')
V = np.zeros((cut*length+1,cut*width+1))
V[0,:] = 10
V[:,0] = 10
V[length*cut,:] = 10
V[:,width*cut] = 10
tem_V = V
fig, ax = plt.subplots(figsize = (10,8))
for n in range(N):
    for i in range(cut*length+1):
        for j in range(cut*width+1):
            if i != 0 and i != cut*length:
                if j != 0 and j != cut*width:
                    tem_V[i][j] = (V[i-1][j]+V[i][j-1]+V[i+1][j]+V[i][j+1])/4
    V = tem_V

cs = ax.contourf(X,Y,V)
fig.colorbar(cs)
```
