## 2D Coulomb potential

The Coulomb potential is logarithmic is two dimensions
$$
  V(\vec{r}) = - \ln\left( \frac{|\vec{r}|}{L} \right)
$$

$$
  \vec{E} = \nabla(V) = \left( \frac{x}{x^2+y^2} , \frac{y}{x^2+y^2} \right)
$$

## Use sympy to verify

```python
from sympy import *

x, y, r, V = symbols('x y r V')
r = x*x+y*y
V = -log(sqrt(x*x+y*y))
V
```

```python
Ex = -diff(V, x)
Ex
```

```python
Ex = -diff(V, y)
Ex
```

## Plot 2D Coulomb potential

```python
N = 50
x = np.linspace(-1, +1, 2*N)
y = np.linspace(-1, +1, 2*N)

X, Y = np.meshgrid(x, y, indexing='xy')
R = np.sqrt(X**2+Y**2)
V = -np.log(R)

fig, ax = plt.subplots(figsize=(12,10))
CS = ax.contourf(X, Y, R)
fig.colorbar(CS)

# ax.plot(X[N,:], Y[N,:], 'ro-')
# ax.plot(X[0,:], V[0,:], 'ko-')
ax.grid()
```
