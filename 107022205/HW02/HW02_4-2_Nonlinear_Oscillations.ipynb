## 4.2 Nonlinear Oscillations

&emsp; In such cases, the $magnitude$ of the force exerted on a particle is the same at $-x$ as at $x$; the $direction$ of the force is opposite in the two cases. Therefore, in a symmetric situation, the first correction to a linear force must be a term proportional to $x^3$; hence,

$$
    F(x) ~= -kx + \varepsilon x^3 \tag{4.4}
$$

where $\varepsilon$ is usually a small quantity. The potential corresponding to such a force is

$$
    U(x) = \frac{1}{2} kx^2 - \frac{1}{4} \varepsilon x^4 \tag{4.5}
$$



```python
import numpy as np
import matplotlib.pyplot as plt

k = 5.5
eps0 = 0
epsS = 0.15
epsH = -0.15

start = -10
end = 10
resolution = (end-start)*100

x = np.linspace(start, end, resolution)

F = -k*x
FS = -k*x + epsS*(x**3)
FH = -k*x + epsH*(x**3)

U = (1/2)*k*(x**2)
US = (1/2)*k*(x**2) - (1/4)*epsS*(x**4)
UH = (1/2)*k*(x**2) - (1/4)*epsH*(x**4)

fig, p = plt.subplots(ncols=2, nrows=2, figsize=(10,10))
p = p.flatten()

p[0].plot(x, F,'--r', lw=3, label='Linear')
p[0].plot(x, FS,'-k', lw=3, label='Soft')
p[0].plot([-5, +5], [0, 0], 'k', lw=1)
p[0].plot([0, 0], [-50, +50], 'k', lw=1)
p[0].set_xlim([-5, 5])
p[0].set_ylim([-50, 50])
p[0].set_xticks([])
p[0].set_yticks([])
p[0].set_xlabel('x')
p[0].set_ylabel('F(x)')
p[0].legend(fontsize=10)

p[1].plot(x, F,'--r', lw=3, label='Linear')
p[1].plot(x, FH,'-k', lw=3, label='Hard')
p[1].plot([-5, +5], [0, 0], 'k', lw=1)
p[1].plot([0, 0], [-50, +50], 'k', lw=1)
p[1].set_xlim([-5, 5])
p[1].set_ylim([-50, 50])
p[1].set_xticks([])
p[1].set_yticks([])
p[1].set_xlabel('x')
p[1].set_ylabel('F(x)')
p[1].legend(fontsize=10)

p[2].plot(x, U,'--r', lw=3, label='Parabolic')
p[2].plot(x, US,'-k', lw=3, label='Soft')
p[2].plot([-10, +10], [0, 0], 'k', lw=1)
p[2].plot([0, 0], [0, 50], 'k', lw=1)
p[2].set_xlim([-5, 5])
p[2].set_ylim([0, 50])
p[2].set_xticks([])
p[2].set_yticks([])
p[2].set_xlabel('x')
p[2].set_ylabel('U(x)')
p[2].legend(fontsize=10)

p[3].plot(x, U,'--r', lw=3, label='Parabolic')
p[3].plot(x, UH,'-k', lw=3, label='Hard')
p[3].plot([-10, +10], [0, 0], 'k', lw=1)
p[3].plot([0, 0], [0, 50], 'k', lw=1)
p[3].set_xlim([-5, 5])
p[3].set_ylim([0, 50])
p[3].set_xticks([])
p[3].set_yticks([])
p[3].set_xlabel('x')
p[3].set_ylabel('U(x)')
p[3].legend(fontsize=10)
```

### EXAMPLE 4.1

Consider a particle of mass m suspended between two identical springs (Figure 4-3). Show that the system is nonlinear. Find the steady-state solution for a driving force $F_0 \cos{\omega t}$.

***Solution*** In real physical situations, we are often concerned with symmetric forces and potentials. But some cases have asymmetric forms. For example,

$$
    F(x) = -kx + \lambda x^2 \tag{4.16}
$$

The potential for which is

$$
    U(x) = \frac{1}{2} kx^2 - \frac{1}{3} \lambda x^3 \tag{4.17}
$$

This case is illustrated in Figure 4-4 for $\lambda<0$; the system is hard for $x>0$ and soft for $x<0$.

```python
import numpy as np
import matplotlib.pyplot as plt

k = 10
Lambda = -1
start = -10
end = 10
resolution = (end-start)*100

x = np.linspace(start, end, resolution)

F = -k*x
FL = -k*x + Lambda*(x**2)

U = (1/2)*k*(x**2)
UL = (1/2)*k*(x**2) - (1/3)*Lambda*(x**3)

fig, p = plt.subplots(ncols=2, nrows=1, figsize=(10,5))
p = p.flatten()

p[0].plot(x, F,'--r', lw=3, label='Linear')
p[0].plot(x, FL,'-k', lw=3)
p[0].plot([-5, +5], [0, 0], 'k', lw=1)
p[0].plot([0, 0], [-50, +50], 'k', lw=1)
p[0].set_xlim([-5, 5])
p[0].set_ylim([-50, 50])
p[0].set_xticks([])
p[0].set_yticks([])
p[0].set_xlabel('x')
p[0].set_ylabel('F(x)')
p[0].legend(fontsize=10)

p[1].plot(x, U,'--r', lw=3, label='Parabolic')
p[1].plot(x, UL,'-k', lw=3)
p[1].plot([-5, +5], [0, 0], 'k', lw=1)
p[1].plot([0, 0], [0, 100], 'k', lw=1)
p[1].set_xlim([-5, 5])
p[1].set_ylim([0, 100])
p[1].set_xticks([])
p[1].set_yticks([])
p[1].set_xlabel('x')
p[1].set_ylabel('U(x)')
p[1].legend(fontsize=10)
```
