## 4.3 Phase Diagrams for Nonlinear Systems

&emsp; The construction of a phase diagram for a nonlinear system may be accomplished by using Equation 2.97:

$$
    \dot{x}(x) \propto \sqrt{E - U(x)}
$$

For example, consider the asymmetric potential shown in Figure 4-5a, which represents a system that is soft for $x<0$ and hard for $x>0$. If no damping occurs, then because $\dot{x}$ is proportional to $\sqrt{E - U(x)}$, the phase diagram must be of the form shown in Figure 4-5b. Three of the oval phase paths are drawn, corresponding to the three values of the total energy indicated by the dotted lines in the potential diagram. For a total energy only slightly greater than that of the minimum of the potential, the oval phase paths approach ellipses.

```python
import numpy as np
import matplotlib.pyplot as plt

k = 1
n = 1
Lambda = -0.02

start  = -26
end  = 19

resolution = 10000
x = np.linspace(start, end, resolution)

E1 = 200
E2 = 125
E3 = 50

U = (1/2)*k*(x**2) - (1/3)*Lambda*(x**3)

xdx1 = n*np.sqrt(E1 - U)
xdx2 = n*np.sqrt(E2 - U)
xdx3 = n*np.sqrt(E3 - U)

fig, p = plt.subplots(ncols=1, nrows=2, figsize=(6,8))
p = p.flatten()

p[0].set_title('Figure 4-6')
p[0].plot(x, U,'-k', lw=3)
p[0].set_xlim([-30, 30])
p[0].set_ylim([0, 250])
p[0].plot([0, 0], [0, 250], 'k', lw=1)
p[0].plot([-30, 30], [E1, E1], 'r', lw=3, label='E1')
p[0].plot([-30, 30], [E2, E2], 'g', lw=3, label='E2')
p[0].plot([-30, 30], [E3, E3], 'b', lw=3, label='E3')
p[0].set_xlabel('x')
p[0].set_ylabel('U')

p[1].plot(x, xdx1,'-r', lw=3, label='E1')
p[1].plot(x, -xdx1,'-r', lw=3)
p[1].plot(x, xdx2,'-g', lw=3, label='E2')
p[1].plot(x, -xdx2,'-g', lw=3)
p[1].plot(x, xdx3,'-b', lw=3, label='E3')
p[1].plot(x, -xdx3,'-b', lw=3)
p[1].plot([-30, +30], [0, 0], 'k', lw=1)
p[1].plot([0, 0], [-20, 20], 'k', lw=1)
p[1].set_xlim([-30, 30])
p[1].set_ylim([-20, 20])
p[1].set_xlabel('x')
p[1].set_ylabel('dx')
p[1].legend()
```

&emsp; In the vicinity of the *maximum* of a potential, a qualitatively different type of motion occurs (Figure 4-6). Here the point $x = 0$ is one of *unstable* equilibrium, because if a particle is at rest at this point, then a slight disturbance will result in locally unbounded motion.

```python
import numpy as np
import matplotlib.pyplot as plt

k = 30
n = 1
Lambda = -0.002

start  = -60
end  = 60

resolution = 10000
x = np.linspace(start, end, resolution)

Ea = 15000
E1 = 5000
E0 = 0
Eb = -7500
E2 = -15000
Ec = -25000

NU = -((1/2)*k*(x**2) - (1/3)*Lambda*(x**3))

xdxa = n*np.sqrt(Ea - NU)
xdx1 = n*np.sqrt(E1 - NU)
xdx0 = n*np.sqrt(E0 - NU)
xdxb = n*np.sqrt(Eb - NU)
xdx2 = n*np.sqrt(E2 - NU)
xdxc = n*np.sqrt(Ec - NU)

fig, p = plt.subplots(ncols=1, nrows=2, figsize=(6,8))
p = p.flatten()

p[0].set_title('Figure 4-6')
p[0].plot(x, NU,'-k', lw=3)
p[0].plot([0, 0], [-20000, 10000], 'k', lw=1)
p[0].plot([-60, 60], [E1, E1], 'r', lw=3, label='E1')
p[0].plot([-60, 60], [E0, E0], 'g', lw=3, label='E0')
p[0].plot([-60, 60], [E2, E2], 'b', lw=3, label='E2')
p[0].set_xlim([-60, 60])
p[0].set_ylim([-20000, 10000])
p[0].set_xlabel('x')
p[0].set_ylabel('U')
p[0].set_xticks([])
p[0].set_yticks([])

p[1].plot(x, xdx1,'-r', lw=3, label='E1')
p[1].plot(x, -xdx1,'-r', lw=3)
p[1].plot(x, xdx0,'-g', lw=3, label='E0')
p[1].plot(x, -xdx0,'-g', lw=3)
p[1].plot(x, xdx2,'-b', lw=3, label='E2')
p[1].plot(x, -xdx2,'-b', lw=3)

p[1].plot(x, xdxa,'-k', lw=1)
p[1].plot(x, -xdxa,'-k', lw=1)
p[1].plot(x, xdxb,'-k', lw=1)
p[1].plot(x, -xdxb,'-k', lw=1)
p[1].plot(x, xdxc,'-k', lw=1)
p[1].plot(x, -xdxc,'-k', lw=1)

p[1].plot([-60, +60], [0, 0], 'k', lw=1)
p[1].plot([0, 0], [-200, 200], 'k', lw=1)
p[1].set_xlim([-60, 60])
p[1].set_ylim([-150, 150])
p[1].set_xlabel('x')
p[1].set_ylabel('dx')
p[1].set_xticks([])
p[1].set_yticks([])
p[1].legend()
```

&emsp; By referring to the phase paths for the potentials shown in Figures 4-5 and 4-6, we can rapidly construct a phase diagram for any arbitrary potential (such as that in Figure 2-14).

&emsp; An important type of nonlinear equation was extensively studied by van der Pol in his investigation of nonlinear oscillations in vacuum tube circuits of early radios. This equation has the form

$$
    \ddot{x} + \mu (x^2 - a^2) \dot{x} + \omega_0^2 x = 0 \tag{4.19}
$$

where $\mu$ is a small, positive parameter. A system described by van der Pol's equation has the following interesting property. If the amplitude $|x|$ exceeds the critical value $|a|$, then the coefficient of $x$ is positive and the system is damped. But if $|x|<|a|$ , then *negative damping* occurs; that is, the amplitude of the motion *increases*. It follows that there must be some amplitude for which the motion neither increases nor decreases with time. Such a curve in the phase plane is called the **limit cycle** (Figure 4-7) and is the attractor for this system. Phase paths outside the limit cycle spiral *inward*, and those inside the limit cycle spiral *outward*. Inasmuch as the limit cycle defines locally bounded motion, we may refer to the
situation it represents as *stable*.

&emsp; Now, let us turn to the numerical calculation of van der Pol's Equation 4.19. In order to make the calculation simpler and to be able to examine the system's motion, we let $a = 1$ and $\omega = 1$ with appropriate units. Equation 4.19 becomes

$$
    \ddot{x} + \mu (x^2 - 1) \dot{x} + x = 0 \tag{4.20}
$$

```python
from scipy import linspace
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

mua, mub, muc = 0.05, 1, 5
x0, v0 = 1, 0
x2, v2 = 3, 0

def vdpa(t, fa):
    xa, ya = fa
    return [ya, mua*(1 - xa**2)*ya - xa]
def vdpa2(t, fa2):
    xa2, ya2 = fa2
    return [ya2, mua*(1 - xa2**2)*ya2 - xa2]

def vdpb(t, fb):
    xb, yb = fb
    return [yb, mub*(1 - xb**2)*yb - xb]
def vdpb2(t, fb2):
    xb2, yb2 = fb2
    return [yb2, mub*(1 - xb2**2)*yb2 - xb2]

def vdpc(t, fc):
    xc, yc = fc
    return [yc, muc*(1 - xc**2)*yc - xc]
def vdpc2(t, fc2):
    xc2, yc2 = fc2
    return [yc2, muc*(1 - xc2**2)*yc2 - xc2]

start, end = 0, 100
resolution = (end-start)*20

t = np.linspace(start, end, resolution)

sola = solve_ivp(vdpa, [start, end], [x0, v0], t_eval=t)
sola2 = solve_ivp(vdpa2, [start, end], [x2, v2], t_eval=t)

solb = solve_ivp(vdpb, [start, end], [x0, v0], t_eval=t)
solb2 = solve_ivp(vdpb2, [start, end], [x2, v2], t_eval=t)

solc = solve_ivp(vdpc, [start, end], [x0, v0], t_eval=t)
solc2 = solve_ivp(vdpc2, [start, end], [x2, v2], t_eval=t)

fig, p = plt.subplots(ncols=2, nrows=3, figsize=(12,18))
p = p.flatten()

p[0].plot(sola.t, sola.y[0] ,'-r', lw=3, label='$\mu = 0.05$')
p[0].plot([-5, +5], [0, 0], 'k', lw=1)
p[0].set_yticks([-2, -1, 0, 1, 2])
p[0].plot([start, end], [0, 0], '-k', lw=1)
p[0].set_xlim([start, end])
p[0].set_xlabel('t')
p[0].set_ylabel('x')
p[0].legend(loc='upper left')

p[1].set_title('Figure 4-7')
p[1].plot(sola.y[0], sola.y[1] ,'-r', lw=1)
p[1].plot(sola2.y[0], sola2.y[1] ,'--r', lw=1)
p[1].plot([-5, 5], [0, 0], '-k', lw=1)
p[1].plot([0, 0], [-5, 5], '-k', lw=1)
p[1].set_xlim([-3, 3])
p[1].set_ylim([-3, 3])
p[1].set_xlabel('x')
p[1].set_ylabel('$\dot{x}$')

p[2].plot(solb.t, solb.y[0] ,'-g', lw=3, label='$\mu = 1.0$')
p[2].set_yticks([-2, -1, 0, 1, 2])
p[2].plot([start, end], [0, 0], '-k', lw=1)
p[2].set_xlim([start, end])
p[2].set_xlabel('t')
p[2].set_ylabel('x')
p[2].legend(loc='upper left')

p[3].set_title('Figure 4-8')
p[3].plot(solb.y[0], solb.y[1] ,'-g', lw=1)
p[3].plot(solb2.y[0], solb2.y[1] ,'--g', lw=1)
p[3].plot([-5, 5], [0, 0], '-k', lw=1)
p[3].plot([0, 0], [-5, 5], '-k', lw=1)
p[3].set_xlim([-3, 3])
p[3].set_ylim([-3, 3])
p[3].set_xlabel('x')
p[3].set_ylabel('$\dot{x}$')

p[4].plot(solc.t, solc.y[0] ,'-b', lw=3, label='$\mu = 5$')
p[4].set_yticks([-2, -1, 0, 1, 2])
p[4].plot([start, end], [0, 0], '-k', lw=1)
p[4].set_xlim([start, end])
p[4].set_xlabel('t')
p[4].set_ylabel('x')
p[4].legend(loc='upper left')

p[5].plot(solc.y[0], solc.y[1] ,'-b', lw=1)
p[5].plot(solc2.y[0], solc2.y[1] ,'--b', lw=1)
p[5].plot([-5, 5], [0, 0], '-k', lw=1)
p[5].plot([0, 0], [-5, 5], '-k', lw=1)
p[5].set_xlim([-3, 3])
p[5].set_ylim([-3, 3])
p[5].set_xlabel('x')
p[5].set_ylabel('$\dot{x}$')
```
