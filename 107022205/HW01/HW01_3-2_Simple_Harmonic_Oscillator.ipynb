## 3.2 Simple Harmonic Oscillator

$$\tag{107022205 陳宥源} $$

&emsp; The equation of motion for the simple harmonic oscillator may be obtained by substituting the Hooke's Law force into the Newtonian equation $F = ma$. Thus

$$
    -kx = m \ddot{x}
$$

If we define

$$
    \omega_0^2 = k/m
$$

Equation 3.3 becomes

$$
    \ddot{x} + \omega_0^2 x = 0
$$

According to the results of Appendix C, the solution of this equation can be expressed in either of the forms

$$ 
    x(t) = A \sin{(\omega_0 t - \delta)} \tag{3.6a}
$$

$$
    x(t) = A \cos{(\omega_0 t - \phi)} \tag{3.6b}
$$

where the phases $\delta$ and $\phi$ differ by $\pi/2$.

```python
import numpy as np
import matplotlib.pyplot as plt

omega0 = 2*np.pi
delta = 0
A = 1

start = 0
end = 5
resolution = (end-start)*100

t = np.linspace(start, end, resolution)
x = A*np.sin(omega0*t - delta)

fig, p = plt.subplots(figsize=(10,5))

p.plot(t, x,'-k', lw=3)

p.set_xlabel('Wave')
p.set_ylabel('Amplitude')
p.set_yticks([-1, 0, 1])
p.set_xlim([0, 5])
p.set_ylim([-1.2, 1.2])
```

&emsp; We can obtain the relationship between the total energy of the oscillator and the amplitude of its motion as follows. Using Equation 3.6a for $x(t)$, we find for the kinetic energy,

$$ 
    \begin{aligned}
        T = \frac{1}{2} m \dot{x}^2 &= \frac{1}{2} m \omega_0^2 A^2 \cos^2{(\omega_0 t - \delta)} \\
                                    &= \frac{1}{2} kA^2 \cos^2{(\omega_0 t - \delta)}  \\
    \end{aligned} \tag{3.7}
$$

&emsp; The potential energy may be obtained by calculating the work required to displace the particle a distance x. The incremental amount of work dWnecessary to move the particle by an amount dx against the restoring force $F$ is

$$
    dW = -F dx = kx dx
$$

Integrating from $0$ to $x$ and setting the work done on the particle equal to the potential energy, we have

$$
    U = \frac{1}{2} k x^2 \tag{3.9}
$$

Then

$$
    U = \frac{1}{2} kA^2 \sin^2{(\omega_0 t - \delta)} \tag{3.10}
$$

Combining the expressions for $T$ and $U$ to find the total energy $E$, we have

$$
    E = T + U = \frac{1}{2} kA^2 [\cos^2{(\omega_0 t - \delta)} + \sin^2{(\omega_0 t - \delta)}]
$$

$$
    E = T + U = \frac{1}{2} kA^2 \tag{3.11}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

omega0 = 2*np.pi
delta = 0
k = 1
A = 1

start = 0
end = 3
resolution = (end-start)*100

t = np.linspace(start, end, resolution)
x = A*np.sin(omega0*t - delta)
T = (1/2)*k*(A**2)*(np.cos(omega0*t - delta))**2
U = (1/2)*k*(A**2)*(np.sin(omega0*t - delta))**2
E = T + U

fig, p = plt.subplots(figsize=(10,5))

p.plot(t, x,'--k', lw=3, label='Wave')
p.plot(t, T,'-r', lw=3, label='Kinetic')
p.plot(t, U,'-g', lw=3, label='Potential')
p.plot(t, E,'-b', lw=3, label='Energy')

p.set_xlabel('Wave')
p.set_ylabel('Amplitude')
p.set_yticks([0, 1, 2, 3])
p.set_yticks([-1, 0, 1])
p.set_xlim([0, 3])
p.set_ylim([-1, 1])
p.legend(fontsize=10)
```

## 3.3 Harmonic Oscillations in Two Dimensions

&emsp; In the general case of two-dimensional oscillations, the angular frequencies for the motions in the $x-$ and $y-$ directions need not be equal, so that Equation 3.19 becomes

$$
    \begin{aligned}
        x(t) &= A \cos{(\omega_x t - \alpha)} \\
        y(t) &= B \cos{(\omega_y t - \beta)}  \\
    \end{aligned} \tag{3.27}
$$

where $\delta = \alpha - \beta$

```python
import numpy as np
import matplotlib.pyplot as plt
count = 0

k = 1
m = 1
omega0 = np.sqrt(k/m)

beta = 0
A = 1
B = 1

delta = np.array([90, 120, 150, 180, 210, 240, 270, 300, 330, 360])/180.*np.pi

fig, p = plt.subplots(ncols=5, nrows=2, figsize=(4*5, 4*2))
p = p.flatten()

for count in range(10):
    alpha = beta + delta[count]
    t = np.linspace(0, 10, 100)
    x = A*np.cos(omega0*t - alpha)
    y = B*np.cos(omega0*t - beta)
    p[count].plot(x, y, '-k', lw=3)
    p[count].plot([-A, +A], [0, 0], 'k')
    p[count].plot([0, 0], [-B, +B], 'k')
    p[count].set_xlim([-A, +A])
    p[count].set_ylim([-B, +B])
```

```python
import numpy as np
import matplotlib.pyplot as plt
count = 1
beta = 0

k = 1
m = 1
omega0 = np.sqrt(k/m)

A = 1
B = 1

p = 5
q = 3

delta = 1.5*np.pi
alpha = beta + delta

start = 0
end = 20
resolution = (end-start)*100

t = np.linspace(start, end, resolution)
x = A*np.cos(p*omega0*t - alpha)
y = B*np.cos(q*omega0*t - beta)

fig, f = plt.subplots(figsize=(5,5))
f.plot(x, y,'-k', lw=3)
f.plot([-1.2, +1.2], [0, 0], 'k', lw=1)
f.plot([0, 0], [-1.2, +1.2], 'k', lw=1)
f.set_xlim([-1.2, +1.2])
f.set_ylim([-1.2, +1.2])
plt.title('FIGURE 3-3 Lissajous curve 5:3')

deltaA = np.array([0,         np.pi/3,   np.pi/2])
betaA  = np.array([1.5*np.pi, 1.5*np.pi, np.pi])
alphaA = betaA + deltaA
pA = 1
qA = 2

fig, fA = plt.subplots(ncols=3, nrows=1, figsize=(5*3, 5))
fA = fA.flatten()

for count in range(3):
    tA = np.linspace(0, 10, 100)
    xA = A*np.cos(pA*omega0*t - alphaA[count])
    yA = B*np.cos(qA*omega0*t - betaA[count])
    fA[count].plot(xA, yA, '-k', lw=3)
    fA[count].plot([-A, +A], [0, 0], 'k', lw=1)
    fA[count].plot([0, 0], [-B, +B], 'k', lw=1)
    fA[count].set_xlim([-A, +A])
    fA[count].set_ylim([-B, +B])
```

&emsp; According to the results of the preceding section, we have, for the simple harmonic oscillator

$$
    x(t) = A \sin{(\omega_0 t - \delta)} \tag{3.28a}
$$

$$
    \dot{x}(t) = A \omega_0 \sin{(\omega_0 t - \delta)} \tag{3.28b}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
count = 1

k = 1
m = 1
omega0 = np.sqrt(k/m)
delta = np.pi/4

start = 0
end = 10
resolution = (end-start)*100
t = np.linspace(start, end, resolution)

fig, p = plt.subplots(figsize=(10,5))
p.plot([-3, +3], [0, 0], 'k', lw=1)
p.plot([0, 0], [-3, +3], 'k', lw=1)

for A in [0.5, 1.0, 1.5, 2.0]:
    x = A*np.sin(omega0*t - delta)
    v = omega0*A*np.cos(omega0*t - delta)
    p.plot(x, v, '-k', lw=3)
    
p.set_xlim([-3, 3])
p.set_ylim([-3, 3])
p.set_xlabel('x')
p.set_ylabel('dx')
```
