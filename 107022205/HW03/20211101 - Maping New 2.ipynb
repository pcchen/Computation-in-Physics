## Introduction

Consider the logic eqatuion:
$$
  f(\alpha,x)=\alpha x(1-x).
$$

The iterative equation becomes
$$
  x_{n+1} = f(\alpha, x_n) = \alpha x_n (1-x_n).
$$

We will assume $x$ is restricted to the domain $(0,1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

def plot_f(start, end, resolution, N, alpha, x0, ax=None):
    t = np.linspace(start, end, resolution)
    h = f(alpha, t)
    
    ax.plot(t, h, '-k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)
    
    x = x0
    ax.plot()
    
    ax.plot([x, x], [0, f(alpha, x)], color = 'r')
    
    for i in range(N):
        y = f(alpha, x)
        
        
        ax.arrow(x, x, 0, (y-x)/1.5, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, x], [x, y], color = 'r')
        
        ax.arrow(x, y, (y-x)/1.5, 0, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, y], [y, y], color = 'r')
        
        x = y
    
    ax.plot([1-(1/alpha), 1-(1/alpha)], [0, 1], '--k')
    ax.plot(1-(1/alpha), f(alpha, 1-(1/alpha)), '.k', markersize = 15)
    ax.set_xlim(0, 1.01)
    ax.set_ylim(0, 1.01)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0', '1'], fontsize=12, fontname='Times New Roman')
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['', '1'], fontsize=12, fontname='Times New Roman')
    ax.tick_params(direction='in', length=10, width=1, colors='k')
    ax.set_xlabel(r'$x_n$', fontsize=14, labelpad=-10, fontname='Times New Roman')
    ax.set_ylabel(r'$x_{n+1}$',labelpad = 10, rotation=0, fontsize=14, fontname='Times New Roman')
    
    text_name = r'$\alpha$ = ' + str('%.2f' % alpha) + "\n" + r'$x_0$ = ' + str('%.2f' % x0)
    
    t = ax.text(0.1, 0.8, text_name, fontsize=14, fontname='Times New Roman')
    t.set_bbox(dict(facecolor='white', alpha=1, edgecolor='none'))
    
fig, (ax01) = plt.subplots(figsize=(6, 6))

plot_f(0, 1, 1000, 20, 2.9, 0.19, ax=ax01)
```

&emsp; Define the function $f(\alpha, x)$

$$
    f(\alpha, x) = \alpha x(1-x)
$$

$$
    \frac{df(\alpha, x)}{dx} = \alpha (1-2x)
$$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

def plot_f(start, end, resolution, N, alpha, x0, ax=None):
    t = np.linspace(start, end, resolution)
    h = f(alpha, t)
    
    ax.plot(t, h, '-k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)
    
    x = x0
    ax.plot()
    
    ax.plot([x, x], [0, f(alpha, x)], color = 'r')
    
    for i in range(N):
        y = f(alpha, x)
        
        ax.arrow(x, x, 0, (y-x)/1.5, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, x], [x, y], color = 'r')
        
        ax.arrow(x, y, (y-x)/1.5, 0, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, y], [y, y], color = 'r')
        
        x = y
    
    ax.plot([1-(1/alpha), 1-(1/alpha)], [0, 1], '--k')
    ax.plot(1-(1/alpha), f(alpha, 1-(1/alpha)), '.k', markersize = 15)
    ax.set_xlim(0, 1.01)
    ax.set_ylim(0, 1.01)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0', '1'], fontsize=12, fontname='Times New Roman')
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['', '1'], fontsize=12, fontname='Times New Roman')
    ax.tick_params(direction='in', length=10, width=1, colors='k')
    ax.set_xlabel(r'$x_n$', fontsize=14, labelpad=-10, fontname='Times New Roman')
    ax.set_ylabel(r'$x_{n+1}$',labelpad = 10, rotation=0, fontsize=14, fontname='Times New Roman')
    
    text_name = r'$\alpha$ = ' + str('%.2f' % alpha) + "\n" + r'$x_0$ = ' + str('%.2f' % x0)
    
    t = ax.text(0.1, 0.8, text_name, fontsize=14, fontname='Times New Roman')
    t.set_bbox(dict(facecolor='white', alpha=1, edgecolor='none'))

    
def plot_df(start, end, resolution, N, alpha, x0, ax=None):
    t = np.linspace(start, end, resolution)
    h = df(alpha, t)
    
    ax.plot(t, abs(h), '-k', lw=2)
    ax.plot([0, 1], [1, 1], '--k', lw=2)
    
    x = x0
    ax.plot()
    
    for i in range(N):
        y = df(alpha, x)
        
        x = y
    
    ax.plot([1-(1/alpha), 1-(1/alpha)], [0, abs(df(alpha, 1-(1/alpha)))], '--r')
    ax.plot(1-(1/alpha), abs(df(alpha, 1-(1/alpha))), '.r', markersize = 15)
    
    ax.set_xlim(0, 1.01)
    ax.set_ylim(0, 4.01)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0', '1'], fontsize=12, fontname='Times New Roman')
    ax.set_yticks([0, 1, 2, 3, 4])
    ax.set_yticklabels(['', '1', '2', '3', '4'], fontsize=12, fontname='Times New Roman')
    ax.tick_params(direction='in', length=10, width=1, colors='k')
    ax.set_xlabel(r'$x_n$', fontsize=14, labelpad=-10, fontname='Times New Roman')
    ax.set_ylabel(r'$\frac{d(x_n)}{dx}$',labelpad = 10, rotation=0, fontsize=14, fontname='Times New Roman')

    
fig, (ax01, ax11, ax21, ax31, ax41) = plt.subplots(ncols=5, nrows=1, figsize=(24, 4))
fig, (ax02, ax12, ax22, ax32, ax42) = plt.subplots(ncols=5, nrows=1, figsize=(24, 4))

plot_f(0, 1, 1000, 20, 2, 0.19, ax=ax01)
plot_f(0, 1, 1000, 20, 2.6, 0.19, ax=ax11)
plot_f(0, 1, 1000, 20, 3.25, 0.19, ax=ax21)
plot_f(0, 1, 1000, 20, 3.47, 0.19, ax=ax31)
plot_f(0, 1, 1000, 20, 4, 0.19, ax=ax41)

plot_df(0, 1, 1000, 20, 2, 0.19, ax=ax02)
plot_df(0, 1, 1000, 20, 2.6, 0.19, ax=ax12)
plot_df(0, 1, 1000, 20, 3.25, 0.19, ax=ax22)
plot_df(0, 1, 1000, 20, 3.47, 0.19, ax=ax32)
plot_df(0, 1, 1000, 20, 4, 0.19, ax=ax42)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

ca = 0  #counter1

fig, ax = plt.subplots(ncols=2, figsize=(12, 5))

alpha_start = 2.999
alpha_end   = 3.6
alpha_total = 4

Start_R, Start_G, Start_B = 1, 0, 0
End_R,   End_G,   End_B   = 0, 1, 0

R_array = np.linspace(Start_R, End_R, alpha_total)
G_array = np.linspace(Start_G, End_G, alpha_total)
B_array = np.linspace(Start_B, End_B, alpha_total)

alpha_array = np.linspace(alpha_start, alpha_end, alpha_total)

iterations = 1000
x0 = 0.5

for alpha in alpha_array:
    
    x = [x0]
    
    for i in range(iterations):
        x.append(f(alpha, x[i-1]))
    
    x = np.array(x)
    
    lable_name = r'$\alpha$ = ' + str('%.3f' % alpha)
    
    ax[0].plot(x, '.', color = [R_array[ca], G_array[ca], B_array[ca]], markersize = 2, label = lable_name)
    ax[1].plot(np.log(np.abs(df(alpha, x[1:]))), '.', color = [R_array[ca], G_array[ca], B_array[ca]], markersize = 2, label = lable_name)

    ax[0].set_xlabel('Iteration times', fontsize=14, labelpad=10, fontname='Times New Roman')
    ax[1].set_xlabel('Iteration times', fontsize=14, labelpad=10, fontname='Times New Roman')
    
    ax[0].set_ylabel(r'$f(\alpha, x)$',labelpad = 5, rotation=90, fontsize=14, fontname='Times New Roman')
    
    ca = ca + 1

L1 = ax[1].legend(loc='lower right', fontsize=12)
plt.setp(L1.texts, family='Times New Roman')
```
