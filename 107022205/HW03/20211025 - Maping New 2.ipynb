## Introduction

Consider the logic eqatuion:
$$
  f(\alpha,x)=\alpha x(1-x).
$$

The iterative equation becomes
$$
  x_{n+1} = f(\alpha, x_n) = \alpha x_n (1-x_n).
$$

We will assume $x$ is restricted to the domain $(0,1)$.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

def plot_f(start, end, resolution, N, alpha, x0, ax=None):
    t = np.linspace(start, end, resolution)
    h = f(alpha, t)
    
    ax.plot(t, h, '-k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)
    
    x = x0
    ax.plot()
    
    ax.plot([x, x], [0, f(alpha, x)], color = 'r')
    
    for i in range(N):
        y = f(alpha, x)
        
        
        ax.arrow(x, x, 0, (y-x)/1.5, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, x], [x, y], color = 'r')
        
        ax.arrow(x, y, (y-x)/1.5, 0, length_includes_head = True, head_width = 0.015, color = 'r')
        ax.plot([x, y], [y, y], color = 'r')
        
        x = y
    
    ax.plot([1-(1/alpha), 1-(1/alpha)], [0, 1], '--k')
    ax.plot(1-(1/alpha), f(alpha, 1-(1/alpha)), '.k', markersize = 15)
    ax.set_xlim(0, 1.01)
    ax.set_ylim(0, 1.01)
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
    ax.set_xticks([0, 1])
    ax.set_xticklabels(['0', '1'], fontsize=12, fontname='Times New Roman')
    ax.set_yticks([0, 1])
    ax.set_yticklabels(['', '1'], fontsize=12, fontname='Times New Roman')
    ax.tick_params(direction='in', length=10, width=1, colors='k')
    ax.set_xlabel(r'$x_n$', fontsize=14, labelpad=-10, fontname='Times New Roman')
    ax.set_ylabel(r'$x_{n+1}$',labelpad = 10, rotation=0, fontsize=14, fontname='Times New Roman')
    
    text_name = r'$\alpha$ = ' + str('%.2f' % alpha) + "\n" + r'$x_0$ = ' + str('%.2f' % x0)
    
    t = ax.text(0.1, 0.8, text_name, fontsize=14, fontname='Times New Roman')
    t.set_bbox(dict(facecolor='white', alpha=1, edgecolor='none'))
    
fig, (ax01) = plt.subplots(figsize=(6, 6))

plot_f(0, 1, 1000, 20, 2.9, 0.19, ax=ax01)
```

### Tasks: Bifucation diagram
* Plot the attractor as a function of $\alpha $ for $0 < \alpha < 4$.

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

ca = 0 #counter a

fig, pB = plt.subplots(figsize=(8, 5), sharex=True)

alpha_start = 0.000001
alpha_end   = 4
alpha_total = 5000

alpha_array = np.linspace(alpha_start, alpha_end, alpha_total)
iterations = 500
last = 50

x0 = 0.1

ly = np.zeros(alpha_total)

x = x0
    
for i in range(iterations):
    
    y = f(alpha_array, x)

    ly = ly + np.log(abs(alpha_array - 2 * alpha_array * x))

    if i >= (iterations - last):
        pB.plot(alpha_array, y, ',k', alpha=.5)
        
    x = y

pB.tick_params(direction='in', length=10, width=1, colors='k', labelsize=12)
pB.set_ylabel(r'$x_n$', loc='center' ,rotation=0, fontsize=14, labelpad=10, fontname='Times New Roman')
pB.set_yticks(np.linspace(0, 1, 11))
pB.set_yticklabels(['0.0','','','','', '0.5','','','','', '1.0'], fontname='Times New Roman')

pB.set_title("Bifurcation diagram", fontsize=16, fontname='Times New Roman')
plt.tight_layout()
```

### Tasks: *three cycle*

Consider the case when $\alpha=3.82831$ and $x_0 = 0.51$. Show that a three cycle occurs with the approximate $x$ values 0.16, 0.52, and 0.96 for the first 80 cycles before the behavior apparently turns chaotic. Find for what iteration that next apparently periodic cycle occurs and for how many cycles it stays periodic?

```python
import numpy as np
import matplotlib.pyplot as plt

def f(alpha, x):
    return alpha*x*(1-x)

def df(alpha, x):
    return alpha*(1-2*x)

fig, ax = plt.subplots(ncols=2, figsize=(12, 5))


alpha = 3.82831
x0 = 0.51
iterations = 500
    
x = [x0]

for i in range(iterations):
    x.append(f(alpha, x[i-1]))
    
x = np.array(x)

lable_name = r'$\alpha$ = ' + str('%.3f' % alpha)

ax[0].plot([0, 500], [0.16, 0.16], '--k', lw=2)
ax[0].annotate(r'$x = 0.16$',
             xy=(400, 0.16), xycoords='data',
             xytext=(+5, +40), textcoords='offset points', fontsize=14,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"), fontname='Times New Roman')

ax[0].plot([0, 500], [0.52, 0.52], '--k', lw=2)
ax[0].annotate(r'$x = 0.52$',
             xy=(400, 0.52), xycoords='data',
             xytext=(+5, +40), textcoords='offset points', fontsize=14,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"), fontname='Times New Roman')

ax[0].plot([0, 500], [0.96, 0.96], '--k', lw=2)
ax[0].annotate(r'$x = 0.96$',
             xy=(200, 0.96), xycoords='data',
             xytext=(-100, -40), textcoords='offset points', fontsize=14,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"), fontname='Times New Roman')

ax[0].plot(x, '.r', markersize = 3, label = lable_name)
ax[0].set_xlabel('Iteration times', fontsize=14, labelpad=10, fontname='Times New Roman')
ax[0].set_ylabel(r'$f(\alpha, x)$',labelpad = 5, rotation=90, fontsize=14, fontname='Times New Roman')
ax[0].set_ylim([0, 1])
ax[0].set_yticks(np.linspace(0, 1, 11))


ax[1].plot(np.log(np.abs(df(alpha, x[1:]))), '.r', markersize = 3, label = lable_name)
ax[1].set_xlabel('Iteration times', fontsize=14, labelpad=10, fontname='Times New Roman')
ax[1].set_ylabel(r'$f(\alpha, x)$' + ' in Log',labelpad = 5, rotation=90, fontsize=14, fontname='Times New Roman')
    
print(np.average(np.log(np.abs(df(alpha, x[1:])))))
```
