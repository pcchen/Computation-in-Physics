```python
import numpy as np
import matplotlib.pyplot as plt
π = np.pi
```

## Equation of motion

By substituting the Hooke's Law $F(x) = -kx$ into the Newtonian equation $F=m\ddot{x}$ we obtain the equation of the motion

$$
  \ddot{x}(t)+\omega^2_o x(t) = 0,
$$

where $\omega^2_0 \equiv k/m$.
The solution is of the form

$$
  x(t) = A \sin(\omega_0 t - \delta),
$$

where $A$ and $\delta$ are determined by the initial conditions.

### Position as a function of time

```python
k = 1.
m = 1.
ω0 = np.sqrt(k/m)

A = 1.
δ = π/4.

t = np.linspace(0, 10)
x = A*np.sin(ω0*t-δ)

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, x)
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.grid()
```

### Kinetic energy, potential energy, and total energy

From the solution $x(t) = A \sin(\omega_0 t - \delta)$ we obtain the velocity $\dot{x}(t) = \omega_0 A \cos(\omega_o t - \delta)$.

The kinetic energy

$$
  T = \frac{1}{2} m \dot{x}^2(t) 
  = \frac{1}{2} m\omega^2_0 A^2 \cos(\omega_0 t - \delta) 
  = \frac{1}{2}k A^2 \cos^2(\omega_0 t - \delta).
$$

The potential energy

$$
  U = \frac{1}{2} k x^2(t)
  = \frac{1}{2} k A^2 \sin^2(\omega_0 t - \delta).
$$

The total energy

$$
  E = T + U = \frac{1}{2} k A^2.
$$

```python
k = 1.
m = 1.
ω0 = np.sqrt(k/m)

A = 1.
δ = π/4.

t = np.linspace(0, 10, 1000)
x = A*np.sin(ω0*t-δ)
v = ω0*A*np.cos(ω0*t-δ)

T = 0.5*m*v*v
U = 0.5*k*x*x
E = T + U

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, T, label='Kinetic energy')
ax.plot(t, U, label='Potential energy')
ax.plot(t, E, 'k--', label='Total energy')

ax.set_xlabel('t')
ax.legend()
```

## Harmonic Oscillations in Two Dimensions

Consider the harmonic oscillations in 2D, with the eqatuion $F_x=-kx$ and $F_y=-ky$. 
The equations of motions are

$$
  \ddot{x}(t)+\omega^2_o x(t) = 0,
$$
$$
  \ddot{y}(t)+\omega^2_o y(t) = 0,
$$

The solutions are

$$
  x(t) = A \cos(\omega_0 t - \alpha)
$$
$$
  y(t) = B \cos(\omega_0 t - \beta)
$$

where $A$, $B$, and $\delta$ are determined by the initial conditions.

### The path of the motion

```python
k = 1.
m = 1.
ω0 = np.sqrt(k/m)
A = 1.
B = 1.

δ = π/4.
β = 0.
α = β + δ

fig, ax = plt.subplots(figsize=(6,6))
t = np.linspace(0, 10)
x = A*np.cos(ω0*t-α)
y = B*np.cos(ω0*t-β)
ax.plot(x, y)
ax.plot([-A, +A], [0, 0], 'k')
ax.plot([0, 0], [-B, +B], 'k')
ax.set_xlim([-A, +A])
ax.set_ylim([-B, +B])
```

```python
k = 1.
m = 1.
ω0 = np.sqrt(k/m)
A = 1.
B = 1.

β = 0.
δ = np.array([90, 120, 150, 180, 210, 240, 270, 300, 330, 360])/180.*np.pi

fig, ax = plt.subplots(ncols=5, nrows=2, figsize=(4*5, 4*2))
ax = ax.flatten()

for i in range(10):
    α = β + δ[i]
    t = np.linspace(0, 10)
    x = A*np.cos(ω0*t-α)
    y = B*np.cos(ω0*t-β)
    ax[i].plot(x, y)
    ax[i].plot([-A, +A], [0, 0], 'k')
    ax[i].plot([0, 0], [-B, +B], 'k')
    ax[i].set_xlim([-A, +A])
    ax[i].set_ylim([-B, +B])
```

### Lissajour curve

In the general case, one has $k_x \neq k_y$ and $\omega_x=\sqrt{k_x/m} \neq \omega_y=\sqrt{k_y/m}$.
The equations of motions become

$$
  \ddot{x}(t)+\omega^2_x x(t) = 0,
$$
$$
  \ddot{y}(t)+\omega^2_y y(t) = 0,
$$

The solutions become

$$
  x(t) = A \cos(\omega_x t - \alpha)
$$
$$
  y(t) = B \cos(\omega_y t - \beta)
$$

where $A$, $B$, and $\delta$ are determined by the initial conditions.

The path of the motion is no longer an ellipse but a **Lissajour curve**.

```python
A = 1.
B = 1.
ωx = 1.
ωy = 2*ωx

β = π*3/4
δ = np.array([0., π/3, π/2])
t = np.linspace(0, 10, 1000)

fig, ax = plt.subplots(ncols=3, figsize=(6*3, 6))
ax = ax.flatten()

for i in range(3):
    α = β + δ[i]
    x = A*np.cos(ωx*t-α)
    y = B*np.cos(ωy*t-β)
    ax[i].plot(x, y)
    ax[i].plot([-A, +A], [0, 0], 'k')
    ax[i].plot([0, 0], [-B, +B], 'k')
    ax[i].set_xlim([-A, +A])
    ax[i].set_ylim([-B, +B])
```

### Phase diagrams

We may consider the quantities $(x(t), \dot{x}(t))$ to be the coordinates of a point in a two-dimensional **phase space**.

```python
k = 1.
m = 1.
ω0 = np.sqrt(k/m)
δ = π/4.
t = np.linspace(0, 10, 1000)

fig, ax = plt.subplots(figsize=(10,8))
for A in [0.5, 1.0, 1.5, 2.0]:
    x = A*np.sin(ω0*t-δ)
    v = ω0*A*np.cos(ω0*t-δ)
    ax.plot(x, v, 'k')

ax.set_xlabel('x', fontsize=24)
ax.set_ylabel('v', fontsize=24)
ax.grid()
```
