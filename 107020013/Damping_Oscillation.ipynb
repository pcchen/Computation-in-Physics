```python
import matplotlib.pyplot as plt
import numpy as np
A=1
Phi=np.pi
m=1
k=1
Wn=np.sqrt(k/m)
t=np.linspace(0,20,101)
fig, ax = plt.subplots(ncols=4,figsize=(20,4))
for i in range(len(ax)):
    c=i
    Zeta=c/2/np.sqrt(k*m)
    Zeta**=2
    if Zeta<=1:
        Wd=Wn*np.sqrt(1-Zeta**2)
    else:
        Wd=Wn*np.sqrt(Zeta**2-1)
    x=A*np.exp(-Zeta*Wn*t)*np.cos(Wd*t+Phi)
    ax[i].plot(t,x,'go-')
    ax[i].set_xlabel('t (s)')
    ax[i].set_ylabel('x (m)')
    title='zeta='+str(Zeta)
    ax[i].set_title(title)
```

**zeta=0** *simple harmonic oscillation*\
**zeta=0.25** *underdamping*\
**zeta=1** *critical damping*\
**zeta=2.25** *overdamping*\
