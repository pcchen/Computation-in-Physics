```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def f(x,alpha):
    return alpha * x * (1-x)

def df(x,alpha):
    return alpha(1-2*x)
```

```python
'''
plot f(x) and y=x
'''

x = np.linspace(0,1,1000)
alpha = np.array([0.8, 1.2, 2.8, 3.2])
y1 = f(x,alpha[0])
y2 = f(x,alpha[1])
y3 = f(x,alpha[2])
y4 = f(x,alpha[3])
y = x
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (10,8));

ax[0,0].plot(x,y1);
ax[0,0].plot(x,y);

ax[0,1].plot(x,y2);
ax[0,1].plot(x,y);

ax[1,0].plot(x,y3);
ax[1,0].plot(x,y);

ax[1,1].plot(x,y4);
ax[1,1].plot(x,y);
```

```python
'''
plot every x versus iteration
'''

x0 = 0.6
x = [[x0], [x0], [x0], [x0]]

alpha = np.array([0.8, 1.2, 2.8, 3.2])

N = 100
for i in range(len(x)):
    for j in range(N-1):
        x[i][j] = f(x[i][j],alpha[i])
        x[i].append(x[i][j])

fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (10,10));
ax[0,0].plot(range(N),x[0],'.');
ax[0,0].set_title('alpha = 0.8')
ax[0,0].set_xlabel('iteration')
ax[0,0].set_ylabel('Xn')
ax[0,1].plot(range(N),x[1],'.');
ax[0,1].set_title('alpha = 1.2')
ax[0,1].set_xlabel('iteration')
ax[0,1].set_ylabel('Xn')
ax[1,0].plot(range(N),x[2],'.');
ax[1,0].set_title('alpha = 2.8')
ax[1,0].set_xlabel('iteration')
ax[1,0].set_ylabel('Xn')
ax[1,1].plot(range(N),x[3],'.');
ax[1,1].set_title('alpha = 3.2')
ax[1,1].set_xlabel('iteration')
ax[1,1].set_ylabel('Xn')
```

```python
'''
plot f(x), y=x, and trajectory
'''

x = np.linspace(0,1,1000)
alpha = np.array([2.8, 3.2, 3.52, 3.54, 4])
y1 = f(x,alpha[0])
y2 = f(x,alpha[1])
y3 = f(x,alpha[2])
y4 = f(x,alpha[3])
y5 = f(x,alpha[4])
y = x

x0 = [[0.6],[0.6],[0.6],[0.6],[0.6]]
N = 100
for i in range(len(x0)):
    for j in range(N-1):
        x0[i][j] = f(x0[i][j],alpha[i])
        x0[i].append(x0[i][j])

fig, ax = plt.subplots(nrows = 5, figsize = (10,40));

ax[0].plot(x,y1);
ax[0].plot(x,y);

ax[1].plot(x,y2);
ax[1].plot(x,y);

ax[2].plot(x,y3);
ax[2].plot(x,y);

ax[3].plot(x,y4);
ax[3].plot(x,y);

ax[4].plot(x,y5);
ax[4].plot(x,y);
```

```python
'''
plot fixpoint versus alpha
'''

x0 = 0.1
alpha = np.linspace(0,4,10000)
N = 1000
for i in range(N):
    x0 = f(x0,alpha)

fig, ax = plt.subplots(figsize=(10,8));
ax.plot(alpha,x0,'.');
```
