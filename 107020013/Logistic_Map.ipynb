```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def f(x,alpha):
    return alpha * x * (1-x)

def df(x,alpha):
    return alpha(1-2*x)
```

```python
'''
plot f(x) and y=x
'''

x = np.linspace(0,1,1000)
alpha = np.array([0.8, 1.2, 2.8, 3.2])
y1 = f(x,alpha[0])
y2 = f(x,alpha[1])
y3 = f(x,alpha[2])
y4 = f(x,alpha[3])
y = x
fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (10,10));

ax[0,0].plot(x,y1);
ax[0,0].plot(x,y);
ax[0,0].set_title('alpha = 0.8')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')

ax[0,1].plot(x,y2);
ax[0,1].plot(x,y);
ax[0,1].set_title('alpha = 0.8')
ax[0,1].set_xlabel('x')
ax[0,1].set_ylabel('y')

ax[1,0].plot(x,y3);
ax[1,0].plot(x,y);
ax[1,0].set_title('alpha = 0.8')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y')

ax[1,1].plot(x,y4);
ax[1,1].plot(x,y);
ax[1,1].set_title('alpha = 0.8')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('y')
```

```python
'''
plot every x versus iteration
'''

x = [[], [], [], []]

alpha = np.array([0.8, 1.2, 2.8, 3.2])

N = 100
for i in range(len(x)):
    x0 = 0.6
    for j in range(N-1):
        x[i].append(x0)
        x0 = f(x0,alpha[i])

fig, ax = plt.subplots(ncols = 2, nrows = 2, figsize = (10,10));
ax[0,0].plot(range(N-1),x[0],'.');
ax[0,0].set_title('alpha = 0.8')
ax[0,0].set_xlabel('iteration')
ax[0,0].set_ylabel('Xn')
ax[0,1].plot(range(N-1),x[1],'.');
ax[0,1].set_title('alpha = 1.2')
ax[0,1].set_xlabel('iteration')
ax[0,1].set_ylabel('Xn')
ax[1,0].plot(range(N-1),x[2],'.');
ax[1,0].set_title('alpha = 2.8')
ax[1,0].set_xlabel('iteration')
ax[1,0].set_ylabel('Xn')
ax[1,1].plot(range(N-1),x[3],'.');
ax[1,1].set_title('alpha = 3.2')
ax[1,1].set_xlabel('iteration')
ax[1,1].set_ylabel('Xn')
```

```python
'''
plot f(x), y=x, and trajectory
'''

x = np.linspace(0,1,1000)
alpha = np.array([2.8, 3.2, 3.52, 3.54, 4])
y1 = f(x,alpha[0])
y2 = f(x,alpha[1])
y3 = f(x,alpha[2])
y4 = f(x,alpha[3])
y5 = f(x,alpha[4])
y = x

x_all = [[],[],[],[],[]]
N = 100
for i in range(len(x_all)):
    x0 = 0.3
    for j in range(N-1):
        x_all[i].append(x0)
        x0 = f(x0,alpha[i])

fig, ax = plt.subplots(nrows = 5, figsize = (10,40));

ax[0].plot(x,y1);
ax[0].plot(x,y);
for i in range(N-2):
    ax[0].plot([x_all[0][i],x_all[0][i]],x_all[0][i:i+2],'g--')
    ax[0].plot(x_all[0][i:i+2],[x_all[0][i+1],x_all[0][i+1]],'g--')

ax[1].plot(x,y2);
ax[1].plot(x,y);
for i in range(N-2):
    ax[1].plot([x_all[1][i],x_all[1][i]],x_all[1][i:i+2],'g--')
    ax[1].plot(x_all[1][i:i+2],[x_all[1][i+1],x_all[1][i+1]],'g--')

ax[2].plot(x,y3);
ax[2].plot(x,y);
for i in range(N-2):
    ax[2].plot([x_all[2][i],x_all[2][i]],x_all[2][i:i+2],'g--')
    ax[2].plot(x_all[2][i:i+2],[x_all[2][i+1],x_all[2][i+1]],'g--')

ax[3].plot(x,y4);
ax[3].plot(x,y);
for i in range(N-2):
    ax[3].plot([x_all[3][i],x_all[3][i]],x_all[3][i:i+2],'g--')
    ax[3].plot(x_all[3][i:i+2],[x_all[3][i+1],x_all[3][i+1]],'g--')

ax[4].plot(x,y5);
ax[4].plot(x,y);
for i in range(N-2):
    ax[4].plot([x_all[4][i],x_all[4][i]],x_all[4][i:i+2],'g--')
    ax[4].plot(x_all[4][i:i+2],[x_all[4][i+1],x_all[4][i+1]],'g--')
```

```python
'''
plot fixpoint versus alpha
'''

x0 = 0.1
x = []
alpha = np.linspace(3.5,4,10000)
N = 10000
for i in range(N):
    x0 = f(x0,alpha)
    if i >9900:
        x.append(x0)

fig, ax = plt.subplots(figsize=(10,8));
for i in range(len(x)):
    ax.plot(alpha,x[i],'m.',markersize = 0.05);
```
