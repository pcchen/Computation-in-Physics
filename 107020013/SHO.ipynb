```python
import numpy as np
import matplotlib.pyplot as plt
```

## Equation of SHO

The solution is of the form

$$
  x(t) = A \sin(\omega_0 t - \delta),
$$

where $\omega^2_0 \equiv k/m$, $A$ and $\delta$ are determined by the initial conditions.

### x(t)

```python
k = 3.
m = 1.
w = np.sqrt(k/m)
A = 1.
d = np.pi/2.
t = np.linspace(0, 20,1000)
x = A*np.sin(w*t-d)
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, x)
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.grid()
```

### Kinetic energy, potential energy, and total energy

The kinetic energy

$$
  T  = \frac{1}{2}k A^2 \cos^2(\omega_0 t - \delta).
$$

The potential energy

$$
  U = \frac{1}{2} k A^2 \sin^2(\omega_0 t - \delta).
$$

The total energy

$$
  E = T + U = \frac{1}{2} k A^2.
$$

```python
k = .5
m = 1.
w = np.sqrt(k/m)
A = 1.
d = np.pi/4.
t = np.linspace(0, 10, 1000)
x = A*np.sin(w*t-d)
v = w*A*np.cos(w*t-d)
T = 0.5*m*v*v
U = 0.5*k*x*x
E = T + U

fig, ax = plt.subplots(figsize=(10,8))
ax.plot(t, T, label='Kinetic energy',color='magenta')
ax.plot(t, U, label='Potential energy',color='c')
ax.plot(t, E, 'k--', label='Total energy')

ax.set_xlabel('t')
ax.set_ylabel('E')
ax.legend()
```

## SHO in Two Dimensions

Consider SHO in 2D, The solutions are

$$
  x(t) = A \cos(\omega_0 t - \alpha)
$$
$$
  y(t) = B \cos(\omega_0 t - \beta)
$$

where $A$, $B$, $\alpha$, and $\beta$ are determined by the initial conditions.

### The path of the motion

```python
k = 1.
m = 1.
w = np.sqrt(k/m)
A = 1.
B = 1.
d = np.pi/4.
b = 0.
a = b + d
t = np.linspace(0, 10)
x = A*np.cos(w*t-a)
y = B*np.cos(w*t-b)
fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y, 'g')
ax.plot([-A, +A], [0, 0], 'k')
ax.plot([0, 0], [-B, +B], 'k')
ax.set_xlim([-A, +A])
ax.set_ylim([-B, +B])
```

### More examples

```python
k = 1.
m = 1.
w = np.sqrt(k/m)
A = 1.
B = 1.
b = 0.
d = np.array([90, 120, 150, 180, 210, 240, 270, 300, 330, 360])/180.*np.pi
fig, ax = plt.subplots(ncols=5, nrows=2, figsize=(4*5, 4*2))
ax = ax.flatten()
for i in range(10):
    a = b + d[i]
    t = np.linspace(0, 10)
    x = A*np.cos(w*t-a)
    y = B*np.cos(w*t-b)
    ax[i].plot(x, y)
    ax[i].plot([-A, +A], [0, 0], 'k')
    ax[i].plot([0, 0], [-B, +B], 'k')
    ax[i].set_xlim([-A, +A])
    ax[i].set_ylim([-B, +B])
```

### Lissajour curve

In the general case, one has $k_x \neq k_y$ and $\omega_x=\sqrt{k_x/m} \neq \omega_y=\sqrt{k_y/m}$.
The equations of motions become

$$
  \ddot{x}(t)+\omega^2_x x(t) = 0,
$$
$$
  \ddot{y}(t)+\omega^2_y y(t) = 0,
$$

The solutions become

$$
  x(t) = A \cos(\omega_x t - \alpha)
$$
$$
  y(t) = B \cos(\omega_y t - \beta)
$$

where $A$, $B$, and $\delta$ are determined by the initial conditions.

The path of the motion is no longer an ellipse but a **Lissajour curve**.
