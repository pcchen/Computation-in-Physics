#### Jacobi and Gauss-Seidel methods

* Study the relative speed of the Jacobi and Gauss-Seidel methods.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,100)
v=np.zeros(100)
v[-1]=1000
v1=np.zeros(100)
v1[-1]=1000
fig ,ax=plt.subplots(1,2,figsize=(8*2,8))
for a in range(10):
    ax[0].plot(x,v,label='interation %d' %(a*100))
    for j in range(100):
        v1=v
        for i in range(len(v)-2):
            v1[i+1]=(v[i]+v[i+2])/2
        v=v1
    
ax[0].legend()
ax[0].set_title('Jacobi method')

x=np.linspace(0,1,100)
v=np.zeros(100)
v[-1]=1000
for a in range(10):
    ax[1].plot(x,v,label='interation %d' %(a*100))
    for j in range(100):
        for i in range(len(v)-2):
            v[i+1]=(v[i]+v[i+2])/2
    
ax[1].set_title('Gauss-Seidel method')
ax[1].legend()
plt.show()
```

### Initial conditions

* Study if the initial guess have a significant impact on overall speed.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,100)
v=np.ones(100)
v[1:99]=500
v[-1]=1000
fig ,ax=plt.subplots(figsize=(8,8))
for a in range(10):
    ax.plot(x,v,label='interation %d' %(a*100))
    for j in range(100):
        for i in range(len(v)-2):
            v[i+1]=(v[i]+v[i+2])/2
ax.legend()
plt.show()
```
