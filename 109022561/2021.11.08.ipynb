## Tasks: Jacobi method and Gauss-Seidel method

* Write a program to compute $V$ in a rectangular domain with Dirichlet boundary conditions. 
* Allow the potentials on all four sides to be specified arbitrarily. 
* The number of grid point can be changed.

#### Jacobi and Gauss-Seidel methods

* Study the relative speed of the Jacobi and Gauss-Seidel methods.

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,100)
v=np.zeros(100)
v[-1]=1000
v1=np.zeros(100)
v1[-1]=1000
fig ,ax=plt.subplots(1,2,figsize=(8*2,8))
for a in range(10):
    ax[0].plot(x,v,label='interation %d' %(a*100))
    for j in range(100):
        v1=v
        for i in range(len(v)-2):
            v1[i+1]=(v[i]+v[i+2])/2
        v=v1
    
ax[0].legend()
ax[0].set_title('Jacobi method')

x=np.linspace(0,1,100)
v=np.zeros(100)
v[-1]=1000
for a in range(10):
    ax[1].plot(x,v,label='interation %d' %(a*100))
    for j in range(100):
        for i in range(len(v)-2):
            v[i+1]=(v[i]+v[i+2])/2
    
ax[1].set_title('Gauss-Seidel method')
ax[1].legend()
plt.show()
```

## Tasks: Contour plots (equipotentials) and electric field 

* Write a program to compute $V$ in a rectangular domain with Dirichlet boundary conditions. 
* Allow the potentials on all four sides to be specified arbitrarily. 
* The number of grid point can be changed.
* Use Jacobi method or Gauss-Seidel method.
* Use some good initial guess.
* Plot the equipotentials.
* Plot the electric fields.

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.full((11,11),0)
x=np.linspace(0,10,11)
y=np.linspace(0,10,11)
X,Y=np.meshgrid(x,y)
data[0,:]=10
data[-1,:]=10
data[:,0]=10
data[:,-1]=10
fig,ax=plt.subplots(figsize=(10,8))
for a in range(10):
    for i in range(data.shape[0]-2):
        for j in range(data.shape[1]-2):
            data[i+1][j+1]=(data[i][j+1]+data[i+2][j+1]+data[i+1][j]+data[i+1][j+2])/4
contour=ax.contourf(X,Y,data,100)
fig.colorbar(contour, shrink=1.0, aspect=20,ax=ax)
Ex, Ey = np.gradient(-data)
ax.quiver(X,Y,Ex,Ey)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
n=21
data1=np.full((n,n),0)
x=np.linspace(-5,5,n)
y=np.linspace(-5,5,n)
X,Y=np.meshgrid(x,y, indexing = 'ij')
data1[0,:]=0
data1[-1,:]=0
data1[:,0]=0
data1[:,-1]=0
fig,ax=plt.subplots(figsize=(10,8))
for a in range(100):
    for i in range(data1.shape[0]-2):
        for j in range(data1.shape[1]-2):
            data1[int(n/2)][int(n/2)]=1000
            data1[i+1][j+1]=(data1[i][j+1]+data1[i+2][j+1]+data1[i+1][j]+data1[i+1][j+2])/4
contour=ax.contourf(X,Y,data1,100)
fig.colorbar(contour, shrink=1.0, aspect=20,ax=ax)
Ex, Ey = np.gradient(-data1)
ax.quiver(X,Y,Ex,Ey)
plt.title("eletric potential and electric field")
plt.show()
fig,ax=plt.subplots(figsize=(8,8))
plt.title("potential at y=0")
ax.plot(x,data1[:,int(n/2)+1],'.')
plt.show()
```
