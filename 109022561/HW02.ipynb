### Solving the pendulum with Euler method
$$
  \frac{d^2 \theta}{dt^2} = \ddot{\theta} = -\omega^2_0 \sin(\theta)
$$
$$
  \theta(t+\delta) = \theta(t) + \dot{\theta}(t) \delta
$$
$$
  \dot{\theta}(t+\delta) = \dot{\theta}(t) + \ddot{\theta}(t) \delta
$$

We set $\theta(0)=2\pi/3$ , $\dot{\theta}(0)=0$ , g=9.8 , L=1 ,$\omega_0^2=g/L$ and $\delta=0.1 ,0.01 ,0.001$

```python
import numpy as np 
import matplotlib.pyplot as plt
g=9.8
L=1
T=2

omega=g/L
dottheta=[[0],[0],[0]]
theta=[[np.pi*2/3],[np.pi*2/3],[np.pi*2/3]]
delta=[0.1,0.01,0.001]
fig,ax= plt.subplots(1,3,figsize=(5*3,5))
for i in range(len(delta)):
    for j in range(int(T/delta[i])):
        theta[i].append(theta[i][j]+dottheta[i][j]*delta[i])
        dottheta[i].append(dottheta[i][j]-omega*np.sin(theta[i][j])*delta[i])
    x=L*np.sin(theta[i])
    y=-L*np.cos(theta[i])
    ax[i].plot(x,y,'.' )
    ax[i].plot(x[0],y[0],'r.')
    ax[i].set_xlim(-L,L)
    ax[i].set_ylim(-L,L)
    ax[i].axhline(y=0, color='k')
    ax[i].axvline(x=0,color='k')
    ax[i].set_title("$\delta$ = %.3f" %delta[i])
plt.savefig("a1.pdf")
plt.show()
```

### Solving the pendulum with Staggered algorithm
$$
  \frac{d^2 \theta}{dt^2} = \ddot{\theta} = -\omega^2_0 \sin(\theta)
$$

$$
  \dot{\theta}(t+dt) = \dot{\theta}(t) + \ddot{\theta}(t) dt
$$
$$
  \theta(t+dt) = x(t) + \dot{\theta}(t+dt) dt
$$

We set $\theta(0)=2\pi/3$ , $\dot{\theta}(0)=0$ , g=9.8 , L=1 ,$\omega_0=g/L$ and $\delta=0.1 ,0.01 ,0.001$

```python
import numpy as np 
import matplotlib.pyplot as plt
g=9.8
L=1
T=1.5

omega=g/L
dottheta=[[0],[0],[0]]
theta=[[np.pi*2/3],[np.pi*2/3],[np.pi*2/3]]
delta=[0.1,0.01,0.001]
fig,ax= plt.subplots(1,3,figsize=(4*3,4))
for i in range(len(delta)):
    for j in range(int(T/delta[i])):
        dottheta[i].append(dottheta[i][j]-omega*np.sin(theta[i][j])*delta[i])
        theta[i].append(theta[i][j]+dottheta[i][j]*delta[i])
    x=L*np.sin(theta[i])
    y=-L*np.cos(theta[i])
    ax[i].plot(x,y,'.' )
    ax[i].plot(x[0],y[0],'r.')
    ax[i].set_xlim(-L,L)
    ax[i].set_ylim(-L,L)
    ax[i].axhline(y=0, color='k')
    ax[i].axvline(x=0, color='k')
    ax[i].set_title("$\delta$ = %.3f" %delta[i])
plt.savefig("a2.pdf")
plt.show()
```

### Solving the pendulum with Velocity Verlet algorithm
$$
  \frac{d^2 \theta}{dt^2} = \ddot{\theta} = -\omega^2_0 \sin(\theta)
$$

$$
  \dot{\theta}(t+dt/2) = \dot{\theta}(t) + \ddot{\theta}(t) dt/2
$$
$$
  \theta(t+dt) = \theta(t) + \dot{\theta}(t+dt/2) dt
$$
$$
  \dot{\theta}(t+dt) = \dot{\theta}(t+dt/2) + \ddot{\theta}(t+dt) dt/2
$$

We set $\theta(0)=2\pi/3$ , $\dot{\theta}(0)=0$ , g=9.8 , L=1 ,$\omega_0=g/L$ and $\delta=0.1 ,0.01 ,0.001$

```python
import numpy as np 
import matplotlib.pyplot as plt
g=9.8
L=1
T=1.5

omega=g/L
dottheta=[[0],[0],[0]]
theta=[[np.pi*2/3],[np.pi*2/3],[np.pi*2/3]]
delta=[0.1,0.01,0.001]
fig,ax= plt.subplots(1,3,figsize=(4*3,4))
for i in range(len(delta)):
    for j in range(int(T/delta[i])):
        a=dottheta[i][j]-omega*np.sin(theta[i][j])*delta[i]/2
        theta[i].append(theta[i][j]+a*delta[i])
        dottheta[i].append(a-omega*np.sin(theta[i][j])*delta[i]/2)
    x=L*np.sin(theta[i])
    y=-L*np.cos(theta[i])
    ax[i].plot(x,y,'.' )
    ax[i].plot(x[0],y[0],'r.')
    ax[i].set_xlim(-L,L)
    ax[i].set_ylim(-L,L)
    ax[i].axhline(y=0, color='k')
    ax[i].axvline(x=0, color='k')
    ax[i].set_title("$\delta$ = %.3f" %delta[i])
plt.savefig("a3.pdf")
plt.show()
```

### Sability of Euler method
T(Total time) =10

g(gravitational)=9.8

L(length)=1

$\theta(0)=0.001$, $\dot{\theta}(0)=0$

I find $\delta_c$=0.2

```python
import numpy as np 
import matplotlib.pyplot as plt
T=10
N=8
g=9.8
L=1
omega=g/L
dottheta=[]
theta=[]
delta=[]
fig,ax= plt.subplots(2,4,figsize=(16,8))
ax = ax.flatten()
for i in range(N):
    dottheta.append([0])
    theta.append([0.001])
    delta.append(0.1*(i+1))
    for j in range(int(T/delta[i])):
        theta[i].append(theta[i][j]+dottheta[i][j]*delta[i])
        dottheta[i].append(dottheta[i][j]-omega*np.sin(theta[i][j])*delta[i])
    x=L*np.sin(theta[i])
    y=-L*np.cos(theta[i])
    ax[i].plot(x,y,'.')
    ax[i].plot(x[0],y[0],'.r',markersize=20)
    ax[i].set_xlim(-L,L)
    ax[i].set_ylim(-L,L)
    ax[i].axhline(y=0, color='k')
    ax[i].axvline(x=0, color='k')
    ax[i].set_title("$\delta$ = %.1f" %delta[i])
plt.savefig("a4.pdf")
plt.show()
```
