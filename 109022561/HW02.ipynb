```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,1,1000)
y=x
α=2
fig ,ax=plt.subplots(figsize=(8/5*6.9,8))
#ax.plot(x,y)
ax.plot(x,α*x*(1-x))
ax.set_xlim(min(x),max(x))
ax.set_ylim(min(y),max(y))
x1=0.15
y1=0
for i in range(5):
    y2=α*x1*(1-x1)
    ax.arrow(x1,y1,0,y2-y1,head_width = 0.02,head_length = 0.02,length_includes_head=1,color='k')
    ax.arrow(x1,y2,-x1,0,head_width = 0.02,head_length = 0.02,length_includes_head=1,color='k')
    x1=y2
ax.set_xlabel(r'$\chi_n$',fontsize=14)
h=ax.set_ylabel(r'$\chi_{n+1}$',fontsize=14)
h.set_rotation(0)
plt.title('FIGURE4-21(a)',fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(2.8,4,1000)

fig ,ax=plt.subplots(figsize=(8/5*9,8))

xlist,ylist=[],[]
for α in x:
    x1=1.3/6.7
    y1=0
    for i in range(1000):
        y2=α*x1*(1-x1)
        y1=y2
        x1=y2
    for i in range(100):
        y2=α*x1*(1-x1)
        y1=y2
        x1=y2
        xlist.append(α)
        ylist.append(y1)
ax.plot(xlist,ylist,'.')
plt.title(r'FIGURE4-23',fontsize=20)
ax.set_xlim(min(xlist),max(xlist))
ax.set_ylim(min(ylist),max(ylist))
h=ax.set_ylabel(r'$\chi_n$',fontsize=14)
h.set_rotation(0)
plt.show()
```
