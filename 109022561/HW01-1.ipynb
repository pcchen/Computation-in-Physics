## Simple Harmonic oscillation


### Equation Of Motion

$$
F=ma=m\ddot{x}=-kx 
$$
$$
m\frac{d^2x}{dt^2}=-kx
$$
k is a constant
$$
x=A\sin(\omega t+\delta)
$$
$\omega^2$ is equal k/m 
$ A\& \delta$為初始條件


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
k=1
m=1
A=1
delta= [0,pi/2]
t=np.linspace(0,10,100)

plt.figure(figsize=(10,8))
omega=np.sqrt(k/m)

x=A*np.sin(omega*t+delta[0])
plt.plot(t,x,label='δ=0')
plt.grid(1)
plt.legend()
plt.title('Simple Hamonic Oscillation')
plt.show()
```

### Kinetic energy, potential energy, and total energy

From the solution $x(t) = A \sin(\omega_0 t - \delta)$ we obtain the velocity $\dot{x}(t) = \omega_0 A \cos(\omega_o t - \delta)$.

The kinetic energy

$$
  T = \frac{1}{2} m \dot{x}^2(t) 
  = \frac{1}{2} m\omega^2_0 A^2 \cos(\omega_0 t - \delta) 
  = \frac{1}{2}k A^2 \cos^2(\omega_0 t - \delta).
$$

The potential energy

$$
  U = \frac{1}{2} k x^2(t)
  = \frac{1}{2} k A^2 \sin^2(\omega_0 t - \delta).
$$

The total energy

$$
  E = T + U = \frac{1}{2} k A^2.
$$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi
k=1
m=1
A=1
delta= [0,pi/2]
t=np.linspace(0,10,100)

plt.figure(figsize=(10,8))
omega=np.sqrt(k/m)

x=A*np.sin(omega*t+delta[0])
v=A*np.cos(omega*t+delta[0])

T=1/2*m*v**2
U=1/2*k*x**2
plt.plot(t,T,label='Kinetic energy')
plt.plot(t,U,label='Potential energy')
plt.plot(t,T+U,'--',label='Total energy')
plt.grid(1)
plt.legend()
plt.title('Simple Hamonic Oscillation')
plt.show()
```

## Damped Oscillation
$$F=ma=m \ddot x=-kx-b \dot x$$

$$ \frac{d^2x}{dt^2}+\frac{b}{m}\frac{dx}{dt}+\frac{k}{m}x=0$$

$$ \ddot x + 2\beta \dot x + \omega ^2_0 x=0 \qquad \beta \equiv \frac{b}{2m} \quad and \quad \omega_0 \equiv \sqrt{k/m} $$

$$ (\frac{d^2}{dt^2}+2\beta \frac{d}{dt}+\omega_0^2)x=0$$ 

$$(\frac{d}{dt}-\frac{-2\beta+\sqrt{4\beta^2-4\omega_0^2}}{2})(\frac{d}{dt}-\frac{-2\beta-\sqrt{4\beta^2-4\omega_0^2}}{2})x=0$$

$$[\frac{d}{dt}-(\beta+\sqrt{\beta^2-\omega_0^2}))(\frac{d}{dt}+(\beta+\sqrt{\beta^2-\omega_0^2})]x=0$$


$$(\frac{dx}{dt}-(\beta+\sqrt{\beta^2-\omega_0^2})x)=0 \Rightarrow x=A\exp^{(-\beta-\sqrt{\beta^2-\omega_0^2}) t}$$
$$(\frac{dx}{dt}-(\beta-\sqrt{\beta^2-\omega_0^2})x)=0 \Rightarrow x=A\exp^{(-\beta+\sqrt{\beta^2-\omega_0^2} )t}$$

$$x(t)=e^{-\beta t}[A_1 exp(\sqrt{\beta^2-\omega^2_0 }t)+A_2 exp(-\sqrt{\beta^2 - \omega^2_0 }t)]=A_3 e^{-\beta t} \cos{(\sqrt{\omega^2_0-\beta^2} t -\delta)}$$

When $\beta^2-\omega^2_0 >0$ , it's overdamping.

When $\beta^2-\omega^2_0 =0$ , it's critical damping.

When $\beta^2-\omega^2_0 <0$ , it's underdamping.



### 二維簡諧運動
$$ \vec{F} = -k \vec{r}$$
$$ F_x=-kr \cos{\theta} =- kx$$ 
$$ F_y=-kr \sin{\theta} =- ky$$
EOM
$$ \ddot{x}+ \omega_0^2 x=0$$
$$ \ddot{y}+ \omega_0^2 y=0$$
The solution is 
$$ x(t)=A \cos{(\omega_0t-\alpha)}$$
$$ y(t)=B \cos{(\omega_0t-\beta)}$$
$A$、$B$、$\alpha$、$\beta$由初始條件決定，$\omega_0 \equiv \sqrt{k/m}$

```python
import numpy as np
import matplotlib.pyplot as plt
a=1
b=1
k=1
m=1
alpha=np.pi/2
beta=0
omega=np.sqrt(k/m)
t=np.linspace(0,10,100)
x=a*np.cos(omega*t-alpha)
y=a*np.cos(omega*t-beta)
fig ,ax =plt.subplots(figsize=(8,8))
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')

ax.set_title('Harmonic Oscillations in Two Dimensions')
plt.show()
```

### 不同相位差的簡諧運動
$$ x(t)=A \cos{\omega_0t+\alpha}$$
$$ y(t)=B \cos{\omega_0t+\beta}$$
$\alpha-\beta$ 不同時會產生不一樣的圖形


```python
import numpy as np
import matplotlib.pyplot as plt
a=1
b=1
k=1
m=10
beta=0
omega=np.sqrt(k/m)
t=np.linspace(0,100,100)
fig ,ax =plt.subplots(2,5,figsize=(4*5,4*2))
ax = ax.flatten()

alpha=np.array([90,120,150,180,210,240,270,300,330,360])
for i in range(len(alpha)):
    x=a*np.cos(omega*t-alpha[i]/180*np.pi)
    y=b*np.cos(omega*t-beta)
    ax[i].plot(x,y)
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title('δ=%d°' %alpha[i])

plt.show()
```

### Lissajous curve
$$ x(t)=A \cos{\omega_xt+\alpha}$$
$$ y(t)=B \cos{\omega_yt+\beta}$$
When $\omega_x \neq \omega_y$ ,it can be maked.
The example is $\omega_x = \frac{3}{4}\omega_y$ and $\alpha = \beta+\frac{\pi}{2}$

```python
import numpy as np
import matplotlib.pyplot as plt
a=1
b=1
k=16
m=1
alpha=np.pi/2
beta=0
omega_x=np.sqrt(k/m)
omega_y=3/4*omega_x
t=np.linspace(0,100,10000)
fig ,ax =plt.subplots(figsize=(8,8))
x=a*np.cos(omega_x*t-alpha)
y=b*np.cos(omega_y*t-beta)
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.title('Lissajour curve')
plt.show()
```

### Lissajour curve with different δ
$$ x(t)=A \cos{\omega_xt+\alpha}$$
$$ y(t)=B \cos{\omega_yt+\beta}$$
$\delta \equiv \alpha - \beta$
We make curve with $\omega_y=2\omega_x$ and $\delta=0,\pi/3 ,\pi/2$ and $\beta=\frac{3}{2}\pi$

```python
import numpy as np
import matplotlib.pyplot as plt
a=1
b=1
k=1
m=1
beta=np.pi/2*3
delta=[np.pi/4,np.pi/3,np.pi/2]
omega_x=np.sqrt(k/m)
omega_y=omega_x*2
t=np.linspace(0,100,10000)
fig ,ax =plt.subplots(1,3,figsize=(4*3,4))
for i in range(len(delta)):
    alpha=beta+delta[i]
    x=a*np.cos(omega_x*t-alpha)
    y=b*np.cos(omega_y*t-beta)
    ax[i].plot(x,y,'-')
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title('δ=%s°' %(delta[i]/np.pi))

plt.show()
```

### Phase Diagrams
Plot a graphic about X and V
$$X=A \sin(\omega_0 t- \delta)$$
$$V=A \omega_0 \cos(\omega_0 t-\delta)$$


```python
import numpy as np
import matplotlib.pyplot as plt
k=1
m=1
alpha=0
omega=np.sqrt(k/m)
t=np.linspace(0,10,100)
fig ,ax =plt.subplots(figsize=(10,8))
for a in [0.5,1,1.5,2]:
    x=a*np.cos(omega*t-alpha)
    v=a*omega*np.sin(omega*t-alpha)
    ax.plot(x,v,'k')
    ax.set_xlabel('x')
    ax.set_ylabel('V')


ax.set_title('X and V diagrams')
plt.grid()
plt.show()
```
