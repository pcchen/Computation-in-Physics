## Damped oscillations

### Damped oscillatiots
$$m \ddot{x}+b\dot{x}+kx=0$$
The general solution is 
$$x(t)=\exp^{-\beta t}[A_1 \exp(\sqrt{\beta^2-\omega^2_0}t)+A_2 \exp(-\sqrt{\beta^2-\omega^2_0 }t)]$$
And $\beta \equiv b/2m$ & $\omega_0=\sqrt{k/m}$
$$$$
Underdamping:$\omega^2_0>\beta^2$ 
$$$$
Critial damping: $\omega^2_0=\beta^2$
$$$$Overdamping:$\omega^2_0<\beta^2$

```python
import numpy as np
import matplotlib.pyplot as plt
k=1/16
b=1
m=1
beta= b/(2*m)
omega=np.sqrt(k/m)
A1,A2=1,1
t=np.linspace(0,10,100)
fig,ax=plt.subplots(figsize=(10,8))

x=np.exp(-beta*t)*(A1*np.exp(np.sqrt(beta**2-omega**2)*t)+A2*np.exp(-1*np.sqrt(beta**2-omega**2)*t))
ax.plot(t,x,'--',label='underdamping')

k=1/4
omega=np.sqrt(k/m)
x=np.exp(-beta*t)*(A1*np.exp(np.sqrt(beta**2-omega**2)*t)+A2*np.exp(-1*np.sqrt(beta**2-omega**2)*t))
ax.plot(t,x,'.',label='critical damping')

k=2
omega=np.sqrt(k/m)
x=np.exp(-beta*t)*(2*np.cos(np.sqrt(-beta**2+omega**2)*t))
ax.plot(t,x,label='overamping')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
ax.set_xlabel('t')
ax.set_ylabel('x(t)')
ax.legend()
plt.title('Damped Oscillation')
plt.show()
        
```

### Underdamped Motion 
the solution of underdamped motion is 
$$x(t) = A\exp^{-\beta t}\cos(\omega_1 t -\delta)$$
$$\omega_1 \equiv \sqrt{\omega_0^2-\beta^2} $$ 
$$x_{en} = \pm A \exp^{-\beta t}$$
Set $\delta=0$ and m=1 , k=1 and $\beta=0.2\omega$

```python
import numpy as np
import matplotlib.pyplot as plt
m=1
k=1
a=1
delta=0
omega=np.sqrt(k/m)
beta=0.2*omega
omega1=np.sqrt(omega**2-beta**2)
t=np.linspace(0,15,10000)
x=a*np.exp(-beta*t)*np.cos(omega1*t-delta)

fig,ax=plt.subplots(figsize=(10,8))
ax.plot(t,x,label='=0.2$_0$')

x_en_up=a*np.exp(-beta*t)
ax.plot(t,x_en_up,'-.',label='A$e^{-t}$')
x_en_down=-a*np.exp(-beta*t)
ax.plot(t,x_en_down,'--',label='-A$e^{-t}$')
beta=0
omega1=np.sqrt(omega**2-beta**2)
x=a*np.exp(-beta*t)*np.cos(omega1*t-delta)
ax.plot(t,x,'--',label='=0')
ax.legend()
ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')
plt.title('FIGURE3-7')
plt.show()
```

### lost energy


$$x(t) = A\exp^{-\beta t}\cos(\omega_1 t -\delta)$$
$$\dot{x}(t)=-A\exp^{-\beta t}[\beta \cos(\omega_1 t-\delta)+\omega_1 \sin(\omega_1 t - \delta)]$$
We plot the total energy and rate of energy loss for the damped oscillatot.

loss energy is 
$$\frac{dE}{dt}$$
and we plot figure as A=1 , $\omega_0=1$, $\beta=0.2$ , $\delta=0$


```python
import numpy as np 
import matplotlib.pyplot as plt
A=1
g=1
m=1
k=1
omega=1
beta=0.2
delta=0
omega1=np.sqrt(omega**2-beta**2)
t=np.linspace(0,10,100)
x=A*np.exp(-beta*t)*np.cos(omega1*t-delta)
dotx=-A*np.exp(-beta*t)*(beta*np.cos(omega1*t-delta)+omega1*np.sin(omega1*t-delta))
E=1/2*m*dotx*dotx+1/2*k*x**2
fig , ax=plt.subplots(2,1,figsize=(4,8))
ax[0].plot(t,E)
ax[0].set_xlim(min(t),max(t))
ax[0].set_ylim(min(E),max(E))
ax[0].set_title('total enery')
ax[0].set_xlabel('t')
ax[0].set_ylabel('E')
dE=[]
for i in range(len(E)-1):
    dE.append((E[i+1]-E[i])/(t[i+1]-t[i]))
ax[1].plot(t[0:len(t)-1],dE)
ax[1].set_xlim(min(t),max(t))
ax[1].set_ylim(min(dE),max(dE))
ax[1].set_title('loss energy')
ax[1].set_xlabel('t')
ax[1].set_ylabel('dE/dt')
ax[1].axhline(y=0, color='k')
ax[1].axvline(x=0, color='k')
ax[0].axhline(y=0, color='k')
ax[0].axvline(x=0, color='k')
plt.show()
```

### Change Coordiante
$$u=\omega_1 x$$
$$w=\beta x +\dot{x}$$
so 
$$ u = \omega_1 A e^{-\beta t}cos(\omega_1 t -\delta)$$
$$ w = -\omega_1 A e^{-\beta t}sin(\omega_1 t -\delta)$$
And set A=1  , $\beta=0.2$ , k= 1 , m =1 , $\delta=\pi/2$

```python
import numpy as np 
import matplotlib.pyplot as plt
A=1
m=1
k=1
omega=1
beta=0.2
delta=np.pi/2
omega1=np.sqrt(omega**2-beta**2)
t=np.linspace(0,1000,100000)
x=A*np.exp(-beta*t)*np.cos(omega1*t-delta)
dotx=-A*np.exp(-beta*t)*(beta*np.cos(omega1*t-delta)+omega1*np.sin(omega1*t-delta))
u=omega1*x
w=beta*x+dotx
fig,ax=plt.subplots(2,1,figsize=(4,8))
ax[0].grid()
ax[0].set_title('w-u diagram')
ax[0].set_xlim(-1,1)
ax[0].set_ylim(-1,1)
ax[0].plot(u,w)
ax[1].set_title('x-$\dot{x}$ diagram')
ax[1].grid()
ax[1].set_xlim(-1,1)
ax[1].set_ylim(-1,1)
ax[1].plot(x,dotx)

plt.show()
```
