## simple harmonic oscillator
<font size=3>
    From Hooke's Law,we can derive the euation of motion of SHO 
    $$  -kx = m\ddot{x} \tag{3.3} $$           
    Where m is mass, k is the spring constant. 
    Define $ ω_0^2 = k/m $ <br>
    So that we get
    $$ \ddot{x}+ω_0^2x = 0  \tag{3.5}$$  
    and the solution is 
    $$ x(t) = Asin(ω_0t-\delta)  \tag{3.6a}$$
    
</font>

```python
import numpy as np
import matplotlib.pyplot as plt
m = 1
k = 1
ω0 = np.sqrt(k/m)
δ = 0

t = np.linspace(0,10,300)
x = np.sin(ω0*t-δ)

plt.figure(figsize=(8,6))
plt.plot(t,x,'k')
plt.plot([0,10],[0,0],'k--')
plt.xlabel("t",fontsize=15)
plt.ylabel("x(t)",fontsize=15)
plt.xlim(0,10)
plt.ylim(-1,1)
```

## The Energy 
<font size=3>
    the kinetic energy T：
    $$
    T = \frac{1}{2}m\dot{x}^2 = \frac{1}{2}mω_0^2A^2cos(ω_0t-\delta)
    $$
    $$
    =\frac{1}{2}kA^2cos^2(ω_0t-\delta)　　　　　　　
    \tag{3.7}
    $$
    the potential energy U：
    $$
    U=\frac{1}{2}kA^2sin^2(ω_0t-\delta) \tag{3.10}
    $$
    so the total energy is：
    $$
    E = T + U = \frac{1}{2}kA^2 \tag{3.11}
    $$
    
    
    
    
    
</font>

```python
m = 1
k = 1
ω0 = np.sqrt(k/m)
δ = 0
A = 1

t = np.linspace(0,10,300)
x = np.sin(ω0*t-δ)
T = 1/2*k*A**2*(np.cos(ω0*t-δ))**2
U = 1/2*k*A**2*(np.sin(ω0*t-δ))**2
E = T + U

plt.figure(figsize=(10,8))
plt.plot(t,x,'k',label = 'potition')
plt.plot(t,T,label = 'Kinetac Energy')
plt.plot(t,U,label = 'Potential Energy')
plt.plot(t,E,'k--',label = 'Total Energy')
plt.legend()

plt.xlabel("t",fontsize=15)
plt.ylabel("x(t) \n E(t)",fontsize=15)
```

## Harmonic Oscillation in Two Dimensions
<font size = 3>
    the general solutions for two-dimentional harmonic oscillations：
    $$ 
    \begin{cases}
    x(t) = Acos(ω_xt-\alpha)\\
    y(t) = Bcos(ω_yt-\beta)\\
    \end{cases}              \tag{3.27}
    $$
    we first focus on the situation that $ω_x=ω_y=ω_0$
    and $\delta = \alpha-\beta$ 
</font>

```python
m = 1
k = 1
ω0 = np.sqrt(k/m)
α = 0
β = 0
δ = α-β
A = 1
B = 1

t = np.linspace(0,10)
fig, ax_list = plt.subplots(nrows=2, ncols=5 , figsize = (20,8))
ax = ax_list.flatten()

for i in range(0,10):
    β=(90+30*i)/180*np.pi
    x = np.cos(ω0*t-α)
    y = np.cos(ω0*t-β)
    ax[i].plot(x,y,'k')
    ax[i].plot([0,0],[-B,B],'k-')
    ax[i].plot([-A,A],[0,0],'k-')
    ax[i].set_xlim([-A,A])
    ax[i].set_ylim([-B,B])
    ax[i].set_title('δ =' + str(90+30*i) + '°')
    ax[i].set_yticks([])
    ax[i].set_xticks([])
```

</font>

## Equation of Motion for Damped Oscillation
<font size=3>
    $$ \ddot{x} + 2\beta\dot{x} + ω_0^2 = 0 $$
    where β is the damping parameter. <br>
    the general solution is
    $$ x(t) = e^{-\beta t}[A_1 exp(\sqrt{\beta^2-ω_0^2}t) + A_2 exp(-\sqrt{\beta^2-ω_0^2}t)] $$
    <br>
    Underdamping：　　$ω_0^2＞\beta^2$<br>
    Critical damping：　$ω_0^2＝\beta^2$<br>
    Overdamping：　　$ω_0^2＜\beta^2$<br>
    <br>
    for Underdamping, the equation of motion can be rewrittern as
    $$  x(t) = Ae^{-\beta t}cos(\sqrt{ω_0^2-\beta^2}t-\delta)  $$
    <br>
    for critical damping,
    $$
    x(t) = e^{-\beta t}[A_1 exp(\sqrt{\beta^2-ω_0^2}t) + A_2 exp(-\sqrt{\beta^2-ω_0^2}t)](A+Bt)
    $$
</font>

```python
m = 1
k = 50
ω0 = np.sqrt(k/m)
A1 = 100   #A1 is for underdamping
A2 = 100
A3 = 0

δ = 0

t = np.linspace(0,3,1000)

# underdamping
beta = 0.2*ω0
x1 = A1*np.exp(-beta*t)*np.cos(np.sqrt(ω0**2-beta**2)*t-δ)

# critical
beta = 1*ω0  
x2 = np.exp(-beta*t)*( A2*np.exp(np.sqrt(beta**2-ω0**2)*t) + A3*np.exp(-np.sqrt(beta**2-ω0**2)*t) )*(1+7*t)

# overdamping
beta = 4*ω0
x3 = np.exp(-beta*t)*( A2*np.exp(np.sqrt(beta**2-ω0**2)*t) + A3*np.exp(-np.sqrt(beta**2-ω0**2)*t) )


plt.figure(figsize=(10,8))
line1 = plt.plot(t,x1,color = 'red')
line2 = plt.plot(t,x2,color = 'blue',ls = '--')
line3 = plt.plot(t,x3,color = 'green',ls = ':')


plt.xlabel("t",fontsize=15)
plt.ylabel("x(t)",fontsize=15)
plt.grid()
plt.legend(['Underdamping','Critical damping','Overdamping'])
```

## the Amplitude of underdamping
<font size = 3>
    the envelope of displacement:
    $$
    x_{en}=\pm Ae^{-\beta t}   \tag{3.41} \\
    $$
    the ratio of amplitude of the oscillation at two successive maxima is
    <br>
    $$
    \frac{A e^{-\beta T}}{A e^{-\beta (T+\tau_1)}} = e^{\beta \tau_1}    \tag{3.42}
    $$

</font>

```python
m = 1
k = 50
ω0 = np.sqrt(k/m)
A1 = 100   #A1 is for underdamping

δ = 0

t = np.linspace(0,2,1000)

beta = 0.2*ω0
x1 = A1*np.exp(-beta*t)*np.cos(np.sqrt(ω0**2-beta**2)*t-δ)

x2 = A1*np.exp(-beta*t)

beta = 0
x3 = A1*np.exp(-beta*t)*np.cos(np.sqrt(ω0**2-beta**2)*t-δ)

plt.figure(figsize=(10,8))

line1 = plt.plot(t,x1,label = 'β = 0.2ω',color = 'r')
line2 = plt.plot(t,x2,label = '$±Ae^{-βt}$',ls = '--',color='k')
line3 = plt.plot(t,-x2,ls = '--',color='k')
line4 = plt.plot(t,x3,label = 'β = 0',ls = ':',color = 'k')
plt.plot([0,2],[0,0],'k-',linewidth = 1)
plt.xlabel('t',fontsize = 20)
plt.ylabel('Amplitude',fontsize = 20)
plt.xlim([0,2])

plt.legend()
```

## the total Energy and the loss rate of Energy
<font size = 3>
    $$
    x(t)= Ae^{-\beta t}cos(ω_1t-\delta)
    $$
    <br>
    $$
    \dot{x}(t) = -Ae^{-\beta t}[\beta cos(ω_1t-\delta)+ω_1sin(ω_1t-\delta)]
    $$
    <br>
    where $ ω_1 = \sqrt{ω_0^2-\beta^2} $
    
<font>

```python
m = 1
k = 10
ω0 = np.sqrt(k/m)
A1 = 1   #A1 is for underdamping

δ = 0

t = np.linspace(0,5,500)

beta = 0.1*ω0
x = A1*np.exp(-beta*t)*np.cos(np.sqrt(ω0**2-beta**2)*t-δ)
v = -A1*np.exp(-beta*t)*( beta*np.cos(np.sqrt(ω0**2-beta**2)*t-δ) + np.sqrt(ω0**2-beta**2)*np.sin(np.sqrt(ω0**2-beta**2)*t-δ) )

T = 1/2*m*v**2
U = 1/2*k*x**2

E = T + U
plt.figure(figsize=(10,8))
ax = plt.axes()

plt.plot(t,E,label = 'Energy')
y = np.empty(500)    #y is Energy defference per unit time
y[0] = 0

for i in range(1,500):
    y[i] = E[i]-E[i-1]
plt.plot(t,y*40,label = 'Energy loss rate')
plt.plot([0,5],[0,0],'k:')
ax.set_xlim([0,5])
ax.set_ylim([-2,5])
ax.legend()
plt.xlabel('t ➝',fontsize = 16)
plt.ylabel('$ dE/dt$ \n$ E $',fontsize = 16)

plt.xticks([0])
plt.yticks([0])
```

## The phase path
<font size = 3>
    <br>
    \begin{cases}
    x(t)= Ae^{-\beta t}cos(ω_1t-\delta) \\
    \dot{x}(t) = -Ae^{-\beta t}[\beta cos(ω_1t-\delta)+ω_1sin(ω_1t-\delta)] \\
    \end{cases}
    <br>
    express $x(t)$ and $\dot{x}(t)$ as
    \begin{cases}
    u = \omega_1Ae^{-\beta t}cos(\omega t-\delta)\\
    w = -\omega_1Ae^{-\beta t}sin(\omega t-\delta)\\
    \end{cases}
    <br>
    where $u = \omega_1x , w = \beta x +\dot{x}$
    
<font>

```python
m = 1
k = 1
ω0 = np.sqrt(k/m)
A1 = 1   #A1 is for underdamping

δ = np.pi/2

t = np.linspace(0,22,500)

beta = 0.2
x = A1*np.exp(-beta*t)*np.cos(np.sqrt(ω0**2-beta**2)*t-δ)
v = -A1*np.exp(-beta*t)*( beta*np.cos(np.sqrt(ω0**2-beta**2)*t-δ) + np.sqrt(ω0**2-beta**2)*np.sin(np.sqrt(ω0**2-beta**2)*t-δ) )

u = np.sqrt(ω0**2-beta**2)*x
w = beta*x + v

fig, axes = plt.subplots(nrows=1, ncols=2 , figsize = (16,6))
ax= axes.flatten()
ax[1].plot(x,v,'k')
ax[1].set_xlim([-1,1])
ax[1].set_ylim([-1,1])
ax[1].grid()
ax[1].set_xlabel('x',fontsize = 15)
ax[1].set_ylabel('$\dot{x}$',fontsize = 15)
ax[1].set_xticks([-1,-0.5,0,0.5,1])
ax[1].set_yticks([-1,-0.5,0,0.5,1])

ax[0].plot(u,w,'k')
ax[0].grid()
ax[0].set_xlim([-1,1])
ax[0].set_ylim([-1,1])
ax[0].set_xlabel('u',fontsize = 15)
ax[0].set_ylabel('w',fontsize = 15)
ax[0].set_xticks([-1,-0.5,0,0.5,1])
ax[0].set_yticks([-1,-0.5,0,0.5,1])
```

## Phase diagram of Overdamped Oscillation
<font size = 3>
    <br>
    $$
    x(t) = e^{-\beta t}[A_1e^{\omega_2 t} +A_2e^{-\omega_2 t} ]      \tag{3.44}  \\
    $$
    where
    $$
    \omega_2 = \sqrt{\beta^2-\omega_0^2}               \tag{3.45}   \\
    $$
    so that the speed is obvious:
    $$
    \dot{x} = \beta e^{-\beta t}[A_1e^{\omega_2 t} +A_2e^{-\omega_2 t} ] + e^{-\beta t}\omega_2 [A_1e^{\omega_2 t} -A_2e^{-\omega_2 t} ]\\
    $$
    asymptotic line:
    $$
    \dot{x} = -(\beta - \omega_2)x
    $$
</font>

```python
m = 1
k = 50
ω0 = np.sqrt(k/m)
β = 1.1*ω0
ω2 = np.sqrt(β**2-ω0**2)
A1 = [1.5,5,3.2,3.5,1.5,-1,-5.5,-3,-1.2]
A2 = [1.5,-3,-0.3,-4,-2.2,-1,3.7,3.5,3.5]
v0 = np.empty(len(A1))

t = np.linspace(0,5,3000)
y = np.linspace(0,3)

plt.figure(figsize=(10,8))
ax = plt.axes()
ax.plot([-3,3],[0,0],'k-',linewidth = 0.8)
ax.plot([0,0],[-40,40],'k-',linewidth = 0.8)

ax.plot(y,-(β-ω2)*y,'k--')
ax.plot(-y,(β-ω2)*y,'k--')
ax.plot(y,-(β+ω2)*y,'k--')
ax.plot(-y,(β+ω2)*y,'k--')

for i in range(len(A1)):
    v0[i] = -β*(A1[i]+A2[i]) + ω2*(A1[i]-A2[i])
    ax.plot(A1[i]+A2[i],v0[i],'ko')
    
    x = np.exp(-β*t)*(A1[i]*np.exp(ω2*t) + A2[i]*np.exp(-ω2*t))
    v = -β*np.exp(-β*t)*(A1[i]*np.exp(ω2*t) + A2[i]*np.exp(-ω2*t)) + np.exp(-β*t)*ω2*(A1[i]*np.exp(ω2*t) - A2[i]*np.exp(-ω2*t))
    ax.plot(x,v,'k')
ax.annotate('$\dot{x}=-(β+ω_2)x$',xy=(-2.7,30),xytext=(-3.5,38),fontsize = 12,arrowprops = dict(facecolor ='k',
                                                                                         shrink = 0.05,
                                                                                         width = 0.01,
                                                                                         headwidth = 3,))
ax.annotate('$\dot{x}=-(β-ω_2)x$',xy=(-2.5,11.5),xytext=(-4,19),fontsize = 12,arrowprops = dict(facecolor ='k',
                                                                                         shrink = 0.05,
                                                                                         width = 0.01,
                                                                                         headwidth = 3))
ax.annotate('Dots represent \n initial values',xy=(-0.47,28),xytext=(0.2,32),fontsize = 12,arrowprops = dict(facecolor ='k',
                                                                                         shrink = 0.05,
                                                                                         width = 0.01,
                                                                                         headwidth = 3))
ax.text(3.2,-0.5,'x',fontsize = 20)
ax.text(-0.07,41,'$\dot{x}$',fontsize = 20)

ax.text(2,12,'Ⅰ',fontsize = 15)
ax.text(2.95,-8.2,'Ⅱ',fontsize = 15)
ax.text(0.6,-28,'Ⅲ',fontsize = 15)
plt.axis('off')
```
